## Stacks: Last In, First Out (LIFO)

Think of a stack like a **stack of plates**:
- You add (push) plates to the TOP
- You remove (pop) plates from the TOP
- The LAST plate you put on is the FIRST one you take off

Real-life examples:
- Browser back button (stack of pages)
- Undo/Redo in editors
- Call stack in recursion


### Building a Stack Class

We'll use a Python list inside our class to store the data, and OOP to organize our methods.

```python
class Stack:
    """A stack data structure (LIFO)."""
    
    def __init__(self):
        self._items = []  # Private attribute!
    
    def push(self, item):
        """Add an item to the top of the stack."""
        self._items.append(item)
    
    def pop(self):
        """Remove and return the top item."""
        if not self.is_empty():
            return self._items.pop()
        return None  # Stack is empty
    
    def peek(self):
        """Look at the top item without removing it."""
        if not self.is_empty():
            return self._items[-1]
        return None
    
    def is_empty(self):
        """Check if the stack is empty."""
        return len(self._items) == 0
    
    def size(self):
        """Return the number of items."""
        return len(self._items)

# Let's try it!
book_stack = Stack()
book_stack.push('Python Basics')
book_stack.push('Algorithms')
book_stack.push('OOP Guide')

print(f'Top book: {book_stack.peek()}')  # OOP Guide
print(f'Removed: {book_stack.pop()}')     # OOP Guide
print(f'Top now: {book_stack.peek()}')   # Algorithms
```

## Queues: First In, First Out (FIFO)

Think of a queue like a **line at a food stall**:
- People join at the BACK (enqueue)
- People leave from the FRONT (dequeue)
- The FIRST person in line is the FIRST person served

Real-life examples:
- Print queue (documents print in order)
- Task scheduling in operating systems
- BFS (Breadth-First Search) in trees/graphs


### Building a Queue Class

Similar to Stack, but we remove from the front instead of the back.

```python
class Queue:
    """A queue data structure (FIFO)."""
    
    def __init__(self):
        self._items = []
    
    def enqueue(self, item):
        """Add an item to the back of the queue."""
        self._items.append(item)
    
    def dequeue(self):
        """Remove and return the front item."""
        if not self.is_empty():
            return self._items.pop(0)  # Remove from front!
        return None
    
    def front(self):
        """Look at the front item."""
        if not self.is_empty():
            return self._items[0]
        return None
    
    def is_empty(self):
        return len(self._items) == 0
    
    def size(self):
        return len(self._items)

# Let's try it!
printer_queue = Queue()
printer_queue.enqueue('Essay.pdf')
printer_queue.enqueue('Report.docx')
printer_queue.enqueue('Photo.jpg')

print(f'Next to print: {printer_queue.front()}')  # Essay.pdf
print(f'Printed: {printer_queue.dequeue()}')       # Essay.pdf
print(f'Next now: {printer_queue.front()}')        # Report.docx
```

## Stack vs Queue Summary

| Feature | Stack | Queue |
|---------|-------|-------|
| Order | LIFO (Last In, First Out) | FIFO (First In, First Out) |
| Add | push (to top) | enqueue (to back) |
| Remove | pop (from top) | dequeue (from front) |
| Real example | Browser back button | Food stall line |
| Uses list method | append / pop() | append / pop(0) |
