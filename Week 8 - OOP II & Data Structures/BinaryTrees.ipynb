## 0. Why Binary Trees?

Imagine a **choose-your-own-adventure** story:

- At each page, you choose:
  - Option A → go to one page
  - Option B → go to another page

This creates a **tree** of choices.

Binary trees show up in:

- Game maps (multiple paths)
- Decision trees (AI, ML)
- Search trees (storing sorted data efficiently)

Today we’ll build a tiny binary tree for a game world.


## 1. Visual: Upside-Down Rooted Tree

A binary tree is like an upside-down tree:

- Top node = **root**
- Each node can have:
  - A **left** child
  - A **right** child
- Some nodes have no children at all (these are **leaves**)


Example:

            Start
          /     \
        Forest Castle
        / \     /  \
    Lake Cave Gate Throne

We'll model this world in Python.

````python
# Smallest building block: a tree node.

class TreeNode:
    def __init__(self, name):
        self.name = name      # location name / question text / etc.
        self.left = None      # left child
        self.right = None     # right child

    def describe(self):
        left_name = self.left.name if self.left is not None else "None"
        right_name = self.right.name if self.right is not None else "None"
        print(f"You are at: {self.name}. Left: {left_name}, Right: {right_name}")

# Let's create a single node and describe it.
start = TreeNode("Start")
start.describe()
````

## 2. Building a Game Map as a Tree

We’ll build this map:

- Root: `"Start"`
  - Left: `"Forest"`
    - Left: `"Lake"`
    - Right: `"Cave"`
  - Right: `"Castle"`
    - Left: `"Gate"`
    - Right: `"Throne"`

Each connection is just an assignment:

```python
start.left = forest
start.right = castle


````python
# Let's code it out!

# Create all the nodes
start = TreeNode("Start")
forest = TreeNode("Forest")
castle = TreeNode("Castle")
lake = TreeNode("Lake")
cave = TreeNode("Cave")
gate = TreeNode("Gate")
throne = TreeNode("Throne")

# Link them up
start.left = forest
start.right = castle

forest.left = lake
forest.right = cave

castle.left = gate
castle.right = throne

# Try describing a few locations
start.describe()
forest.describe()
castle.describe()
lake.describe()
````

## 3. Walking Along a Path

We can move through the tree by choosing:

- `.left` (maybe "Option A")
- `.right` (maybe "Option B")

Example path:

- Start → left → right = `Start → Forest → Cave`

Let’s simulate a journey.


````python
current = start
current.describe()

print("\nGo left to the Forest.")
current = current.left
current.describe()

print("\nGo right to the Cave.")
current = current.right
current.describe()
````

## 4. Simple Interactive Choices

We can let the **user** choose where to go:

- Type `"L"` to go left
- Type `"R"` to go right
- Type `"Q"` to quit

This makes the tree feel like a mini text adventure.


````python
def play_game(start_node):
    current = start_node
    print("Welcome to the Binary Tree Adventure!")
    print("Type L to go left, R to go right, Q to quit.\n")

    while True:
        current.describe()
        choice = input("Your choice (L/R/Q): ").strip().upper()

        if choice == "Q":
            print("You decided to stop exploring.")
            break
        elif choice == "L":
            if current.left is not None:
                current = current.left
            else:
                print("There is nothing on the left. Try again.\n")
        elif choice == "R":
            if current.right is not None:
                current = current.right
            else:
                print("There is nothing on the right. Try again.\n")
        else:
            print("Invalid choice. Please type L, R, or Q.\n")

# Uncomment the line below to try during class (works best when run live):
# play_game(start)
````

## 5. Tree Traversals: Visiting All Nodes

Visiting every node in a tree is called a **traversal**.

Three classic ways (for binary trees):

1. **Preorder**:  
   Visit current → traverse left → traverse right
2. **Inorder**:  
   Traverse left → visit current → traverse right
3. **Postorder**:  
   Traverse left → traverse right → visit current

We’ll implement preorder and inorder to show how recursion makes this very short and clean.


````python
def print_preorder(node):
    """Visit current node, then left, then right."""
    if node is None:
        return
    print(node.name)
    print_preorder(node.left)
    print_preorder(node.right)

print("Preorder traversal from Start:")
print_preorder(start)
````

````python
def print_inorder(node):
    """Visit left, then current, then right."""
    if node is None:
        return
    print_inorder(node.left)
    print(node.name)
    print_inorder(node.right)

print("Inorder traversal from Start:")
print_inorder(start)
````

## 6. Traversals in Real Life

If you imagine visiting rooms in a game:

- Preorder:
  - You log a room **as soon as you enter it**
- Inorder:
  - You always fully explore the **left side** first, then log the current room, then the right side
- Postorder:
  - You only log a room **after** exploring everything below it

The key idea: a tree gives you **structure**, and recursion lets you
“follow the structure” with very little code.


## 7. (Optional) Tree Height

We can ask: “How tall is this tree?”

**Height** of a node:

- A leaf has height 0 (no children)
- Otherwise: `1 + max(height(left), height(right))`

This is a good excuse to practice recursion one more time.


````python
def height(node):
    if node is None:
        return -1  # empty tree has height -1, leaf will be 0
    return 1 + max(height(node.left), height(node.right))

print("Height of the tree rooted at Start:", height(start))
print("Height of the tree rooted at Forest:", height(forest))
print("Height of a leaf (Lake):", height(lake))
````
