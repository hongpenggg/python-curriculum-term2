# Object-Oriented Programming (OOP)

Welcome to one of the biggest milestones in coding! Up until now, we’ve written lists of instructions. With **OOP**, we build virtual "things" that have their own data and behaviors.

Think of it like a video game: instead of just writing code for "health" and "speed," you create a **Player** object that carries its own health, speed, and weapons wherever it goes.

An **object** is a self-contained unit that bundles together data (what it is) and behavior (what it can do)

## 🏗️ Classes vs. Objects

- **Class (The Blueprint)**: This is the design. It says all players should have a name and a level, but it isn't a specific person yet.
- **Object (The Instance)**: This is the actual character you play as. If the class is the "Cookie Cutter," the object is the "Cookie."

### 🏷️ Attributes and Verbs
- **Attributes (Adjectives)**: Variables that belong to the object (e.g., `hp = 100`).
- **Methods (Verbs)**: Functions that belong to the object (e.g., `.jump()` or `.attack()`).

## 🛠️ Building Your First Blueprint

To create a class, we use the `class` keyword. We also use a special function called `__init__` (the **Constructor**) to set up the starting stats of our object.

A **Constructor** is a special method in a class that automatically runs when you create a new object (an instance of that class)

*Pro Tip: The `self` keyword refers to the specific object we are currently working with!*

```python
# Define a Classroom blueprint for our class!
# Our class has a teacher and multiple students

class Classroom:
    def __init__(self, teacher):
        self.teacher = teacher
        self.students = []

class1 = Classroom("Bob")
```

## 🗂️ Class Attributes vs. Instance Attributes

Not all attributes are the same! There are **two kinds**:

| | Class Attribute | Instance Attribute |
| :--- | :--- | :--- |
| **Where defined** | Inside the class, but **outside** `__init__` | Inside `__init__`, attached to `self` |
| **Shared?** | ✅ Shared across **all** objects of that class | ❌ Unique to **each** object |
| **Access** | `ClassName.attr` or `self.attr` | `self.attr` only |
| **Analogy** | School rules that apply to every class | Each student's personal notebook |

### Class Attributes
Think of these as **school-wide rules** — every classroom in the school follows them. They are defined at the top of the class body.

### Instance Attributes
These are **personal to each object** — just like how each classroom has its own teacher and student list.

```python
class BitByBitClass:
    # CLASS ATTRIBUTE — shared by ALL instances
    school_name = "Bit By Bit Coding"
    max_students = 30

    def __init__(self, teacher_name):
        # INSTANCE ATTRIBUTES — unique to each object
        self.teacher = teacher_name
        self.students = []

# Create two separate classrooms
class_a = BitByBitClass("Mr. Lim")
class_b = BitByBitClass("Mrs. Tan")

# Class attributes are the SAME for both
print(class_a.school_name)   # Bit By Bit Academy
print(class_b.school_name)   # Bit By Bit Academy

# Instance attributes are DIFFERENT for each
print(class_a.teacher)       # Mr. Lim
print(class_b.teacher)       # Mrs. Tan
```

### ⚠️ Watch Out — Changing Class Attributes!

Because class attributes are **shared**, changing one via the class name affects **every** object. But if you change it on a single instance using `self`, Python creates a **new instance attribute** that shadows the class one — only for that object.

This is a very common source of bugs, so pay attention!

```python
# Change via the CLASS — affects ALL instances
BitByBitClass.school_name = "Bit By Bit Global Academy"
print(class_a.school_name)   # Bit By Bit Global Academy
print(class_b.school_name)   # Bit By Bit Global Academy

# Change via an INSTANCE — only affects that one object
class_a.school_name = "Special Branch"
print(class_a.school_name)   # Special Branch  (instance attribute shadows the class one)
print(class_b.school_name)   # Bit By Bit Global Academy  (class attribute unchanged)
```

## Private Attributes

A private attribute is a class variable intended to be hidden from outside access to ensure encapsulation.  

In Python, there is no such thing as a truly **"private"** attribute that is completely locked away from the outside world.

To enforce privacy, Python uses two distinct approaches: a polite warning (internal convention) and a structural speed bump (name mangling). We will only be discussing internal convention. You can look up name mangling if you are interested!

---



### Internal Convention: The Single Underscore (_variable)
When you prefix an attribute or method with a single underscore, you are using a visual convention. It tells other developers, "This is internal machinery. Change it at your own risk."

**How it works**: It is a purely stylistic hint. Python does absolutely nothing behind the scenes to stop someone from reading or modifying it.

**When to use it**: For standard internal variables or helper methods that don't need strict protection but shouldn't be part of your public class.


```python
class DigitalWallet:
    def __init__(self, amount):
        self._balance = amount  # Convention: Please treat this as private!

# We can still access ._balance
```

## The Three "Pillars" of OOP

1. **Encapsulation**: The process of bundling data (attributes) and the methods that operate on that data into a **single unit**, or class. It acts as a protective shield, preventing outside code from accessing an object's internal state directly and requiring them to use public interfaces instead.

2. **Inheritance**: A mechanism that allows a new class (subclass) to acquire the properties and behaviors of an **existing** class (superclass). This creates a hierarchical relationship that promotes **code reusability** and reduces redundancy.

3. **Polymorphism**: The ability of different objects to respond to the **same** message or method call in their own way. For example, a single `speak()` method might cause a "Dog" object to bark and a "Cat" object to meow, allowing a program to treat diverse objects uniformly through a common interface.
