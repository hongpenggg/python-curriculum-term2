Create a `BankAccount` class where the balance is private. Users should only be able to deposit, withdraw, and check balance through defined methods. Withdrawals beyond the available balance should be rejected.

```python
class BankAccount:
    def __init__(self, owner, initial_balance):
        self.owner = owner
        # Trigger name mangling to protect the balance from accidental changes
        self.__balance = initial_balance

    def deposit(self, amount):
        if amount > 0:
            self.__balance += amount
            print(f"Deposited ${amount}. New balance: ${self.__balance}")
        else:
            print("Deposit amount must be positive!")

    def withdraw(self, amount):
        if 0 < amount <= self.__balance:
            self.__balance -= amount
            print(f"Withdrew ${amount}. Remaining balance: ${self.__balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds!")

    def get_balance(self):
        # A public getter method is the proper way to view the private balance
        return self.__balance
```

```python
# Test your code
acc = BankAccount("Alice", 500)
acc.deposit(200)       # Deposited $200. New balance: $700
acc.withdraw(100)      # Withdrew $100. Remaining balance: $600
acc.withdraw(9999)     # Invalid withdrawal amount or insufficient funds!
print(acc.get_balance()) # 600
```

Create a base Animal class with `name` and `sound` attributes and a `speak()` method. Then create `Dog` and `Cat` subclasses. Dogs should also have a `fetch()` method and Cats a `purr()` method. All animals should inherit `speak()`.



```python
# Create an Animal class
class Animal:
    def __init__(self, name, sound):
        self.name = name
        self.sound = sound

    def speak(self):
        print(f"I am a talking animal, my name is {self.name}. {self.sound}!")


# Create a Dog class that inherits from Animal
class Dog(Animal):
    def __init__(self, name):
        # Call the parent constructor and pass the specific sound for a Dog
        super().__init__(name, "Woof")

    def fetch(self, item):
        print(f"{self.name} fetched {item}!")


# Create a Cat class that inherits from Animal
class Cat(Animal):
    def __init__(self, name):
        # Call the parent constructor and pass the specific sound for a Cat
        super().__init__(name, "Meow")

    def purr(self):
        print(f"Purr... I am {self.name}")
```

```python
d = Dog("Rex")
d.speak()    # I am a talking animal, my name is Rex. Woof!
d.fetch("ball")

c = Cat("Luna")
c.speak()    # I am a talking animal, my name is Luna. Meow!
c.purr()
```

Build a Notification base class with a `send(message) `method. Create Email, SMS, and PushNotification subclasses that each implement `send()` differently. Write a `notify_all(notifiers, message)` function that sends the message through every channel.

```python
# Keep Notification class unchanged
class Notification:
    def send(self, message):
        raise NotImplementedError

class Email(Notification):
    def __init__(self, address):
        self.address = address

    def send(self, message):
        print(f"'{message}' emailed to {self.address}")

class SMS(Notification):
    def __init__(self, number):
        self.number = number

    def send(self, message):
        print(f"'{message}' SMS to {self.number}")

class PushNotification(Notification):
    def __init__(self, device_id):
        self.device_id = device_id

    def send(self, message):
        print(f"'{message}' pushed to {self.device_id}")


def notify_all(notifiers, message):
    for notifier in notifiers:
        notifier.send(message)



channels = [
    Email("user@example.com"),
    SMS("+65 9123 4567"),
    PushNotification("device-abc")
]

notify_all(channels, "Your order has shipped!")
```
