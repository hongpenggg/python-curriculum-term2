```python
class Student:
    def __init__(self, name):
        self.name = name
        self._grades = []

    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self._grades.append(grade)
        else:
            print('Grade must be between 0 and 100!')

    def average(self):
        if len(self._grades) == 0:
            return None

        return sum(self._grades) / len(self._grades)

    def letter_grade(self):
        average = self.average()
        if average is None:
            print('Student has no grades yet!')
            return
        if average >= 90:
            letter = 'A'
        elif average >= 80:
            letter = 'B'
        elif average >= 70:
            letter = 'C'
        elif average >= 60:
            letter = 'D'
        else:
            letter = 'F'
        return letter

s1 = Student('Bob')
s1.add_grade(50)
s1.add_grade(100)
print(s1.average())
print(s1.letter_grade())
```

## 🧬 Inheritance: Leveling Up

What if you want a special type of classroom, like a **ProjectGroup**?

Instead of rewriting all the code for teachers and students, we can **Inherit** from our original class. This makes the `ProjectGroup` a "Child" and the `BitByBitClass` the "Parent."

- Use `super().__init__()` to grab everything from the parent's blueprint.

*Pro Tip: The parent can't access any attributes or methods from the child!*

```python
class ProjectGroup(BitByBitClass):
    def __init__(self, teacher_name, project_name):
        # Inherit the teacher and student logic from the parent
        super().__init__(teacher_name)
        self.project_name = project_name

    def display_students(self):
        print(f'Project Name: {self.project_name}')
        super().display_students()


# Create a project group
coding_squad = ProjectGroup("Mrs. Tan", "Build a Game")
print(coding_squad.teacher)
print(coding_squad.school_name)
print(coding_squad.display_students())
```

## The Three "Pillars" of OOP

1. **Encapsulation**: The process of bundling data (attributes) and the methods that operate on that data into a **single unit**, or class. It acts as a protective shield, preventing outside code from accessing an object's internal state directly and requiring them to use public interfaces instead.

2. **Inheritance**: A mechanism that allows a new class (subclass) to acquire the properties and behaviors of an **existing** class (superclass). This creates a hierarchical relationship that promotes **code reusability** and reduces redundancy.

3. **Polymorphism**: The ability of different objects to respond to the **same** message or method call in their own way. For example, a single `speak()` method might cause a "Dog" object to bark and a "Cat" object to meow, allowing a program to treat diverse objects uniformly through a common interface.
