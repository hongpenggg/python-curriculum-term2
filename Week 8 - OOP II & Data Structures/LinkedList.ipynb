## 0. Why Do Linked Lists Exist?

Python already has lists. So why invent another list?

Imagine you are running a **queue** at a bubble tea shop:

- People join at the back (tail)
- People leave from the front (head)
- The queue grows and shrinks all the time

With a **linked list**, we can:

- Add to the back without shifting everyone
- Remove from the front without moving everything down
- Re-arrange “people” by just changing links, not copying data

Today, we’ll build a tiny version of this idea in Python.


## 1. Visual: Beads on a String

Think of a linked list as **beads on a string**:

- Each bead = a **node**
- The string from one bead to the next = the **link** (`next` pointer)

Example:

`[HEAD]  A  ->  B  ->  C  ->  None`

We will build this ourselves using a custom `Node` class.


````python
# Let's start with the smallest building block: a Node.

class Node:
    def __init__(self, value):
        self.value = value   # data inside the bead
        self.next = None     # link to the next bead (node)

# Create a few loose nodes
a = Node("A")
b = Node("B")
c = Node("C")

a.next = b   # A points to B
b.next = c   # B points to C

# Let's walk through the nodes starting from 'a'.
current = a
while current is not None:
    print("Node value:", current.value)
    current = current.next
````

## 2. Understanding `next` (The Pointer)

The `next` attribute is like an arrow:

- `a.next = b` means: from `a` you can **go to** `b`
- If `next` is `None`, there is no next node. It’s the **end**.

We never “see” the arrows directly, but we *follow* them in code:

## 3. Building a LinkedList Class

Typing `a = Node(...)`, `a.next = ...` every time is annoying.

Let’s build a `LinkedList` class that:

- Remembers only the **head** node
- Can:
  - Add a node to the **end** (`append`)
  - Print all nodes (`print_list`)
  - Count how many nodes there are (`length`)


````python
class LinkedList:
    def __init__(self):
        self.head = None  # start with an empty list

    def append(self, value):
        """Add a new node with this value to the end of the list."""
        new_node = Node(value)
        if self.head is None:
            # List is empty, new node becomes the head
            self.head = new_node
        else:
            # Walk to the end, then attach
            current = self.head
            while current.next is not None:
                current = current.next
            current.next = new_node

    def print_list(self):
        current = self.head
        while current is not None:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

    def length(self):
        count = 0
        current = self.head
        while current is not None:
            count += 1
            current = current.next
        return count

# Demo
ll = LinkedList()
ll.append("A")
ll.append("B")
ll.append("C")

ll.print_list()
print("Length:", ll.length())
````

## 4. Linked List vs Python List

Python list:

- Stores everything in **one block** of memory (conceptually)
- Fast random access: `my_list[0]`, `my_list[100]`

Linked list:

- Stores each value in a separate **node**
- Each node knows only:
  - its own value
  - where the next node is

Fast operations:

- Adding/removing at the **front** (head)
- Adding at the **back** (if we keep a tail pointer)

Slower operations:

- Jumping to the 100th element (we must traverse from the head)


## 5. Turning the Linked List into an Action Queue

Let’s imagine a **character** in a game with a list of planned actions:

- `"move_left"`
- `"jump"`
- `"attack"`
- `"heal"`

We want to:

- Add actions at the **back** (`enqueue`)
- Execute actions from the **front** (`dequeue`)

We can reuse our `LinkedList` idea, but give it a more game-like name:
`ActionQueue`.


````python
class ActionQueue:
    def __init__(self):
        self.head = None  # front of the queue
        self.tail = None  # back of the queue (for fast enqueue)

    def enqueue(self, action):
        """Add an action to the back of the queue."""
        new_node = Node(action)
        if self.head is None:
            # empty queue
            self.head = self.tail = new_node
        else:
            # attach to the tail
            self.tail.next = new_node
            self.tail = new_node

    def dequeue(self):
        """Remove and return the action at the front of the queue."""
        if self.head is None:
            return None  # nothing to do
        removed = self.head
        self.head = self.head.next
        # If the queue became empty, reset tail as well
        if self.head is None:
            self.tail = None
        return removed.value

    def is_empty(self):
        return self.head is None

    def show(self):
        current = self.head
        print("ActionQueue:", end=" ")
        while current is not None:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

# Demo
queue = ActionQueue()
queue.enqueue("move_left")
queue.enqueue("jump")
queue.enqueue("attack")

queue.show()

print("Next action:", queue.dequeue())
queue.show()

print("Next action:", queue.dequeue())
print("Next action:", queue.dequeue())
print("Next action:", queue.dequeue())  # should be None
queue.show()
````

## 6. Adding the Queue to a Character

Now let’s connect this to a very simple `Character` class:

- Each character has:
  - a `name`
  - an `ActionQueue`

The character can:

- `plan_action(action)` → adds to the queue
- `perform_next_action()` → dequeues and prints what they do

This is a nice way to show students how:

> **OOP + data structures = “brains” for your objects.**


````python
class Character:
    def __init__(self, name):
        self.name = name
        self.actions = ActionQueue()

    def plan_action(self, action):
        print(f"{self.name} plans to: {action}")
        self.actions.enqueue(action)

    def perform_next_action(self):
        action = self.actions.dequeue()
        if action is None:
            print(f"{self.name} has nothing to do.")
        else:
            print(f"{self.name} performs: {action}")

    def show_plan(self):
        print(f"{self.name}'s current action plan:")
        self.actions.show()

# Demo
hero = Character("Aria")

hero.plan_action("move_left")
hero.plan_action("jump")
hero.plan_action("attack")
hero.show_plan()

hero.perform_next_action()
hero.perform_next_action()
hero.show_plan()

hero.perform_next_action()
hero.perform_next_action()
hero.show_plan()
````
