# Collaborative Challenge: Character status tracker

You need to create a tracking system for a game that can track and update the health of a player.

# Requirements
1. `take_damage(health, damage)`, this function takes the current health of the plater and subtracts the damage value from the health value. Return the updated health. *MAKE SURE THE FUNCTION DOES NOT RETURN NEGATIVE VALUES*
2. `level_up(experience, level, level_up_threshold)`, this function updates the level whenever the experience exceeds the *level_up_threshold*. Return the updated level
3. `apply_poison(health, poison_damage, turns)` recursively applies poison damage each turn until either turns run out, or health falls below 0. Return the health final health.

4. `draw_health_bar(health, max_health)` this function uses the *turtle* library to draw a statis health bar for the player.

#Preset values
- `level_up_threshold = 150`
- `if health > 50%` health bar should be green
- `if health > 20%` health bar should be orange
- `if health < 20%` health bar should be red

*Hint: You can draw two rectangles to represent the health bar!*

```python
import ColabTurtlePlus.Turtle as turtle

def take_damage(health, damage):
    """Subtract damage from health, ensuring health never drops below 0."""
    health -= damage
    return max(0, health)

def level_up(experience, level, level_up_threshold):
    """Increment level by 1 if experience exceeds the threshold."""
    if experience >= level_up_threshold:
        level += 1
        print(f"Level up! You are now level {level}.")
    return level

def apply_poison(health, poison_damage, turns):
    """Recursively apply poison damage each turn until turns run out or health reaches 0."""
    if turns == 0:
        print("Poison has worn off!")
        return health
    if health <= 0:
        print("Player has died from poison!")
        return 0

    health = take_damage(health, poison_damage)
    print(f"Poison deals {poison_damage} damage! HP remaining: {health} | Turns left: {turns - 1}")
    return apply_poison(health, poison_damage, turns - 1)

def draw_health_bar(health, max_health):
    """Draw a health bar using turtle, coloured based on remaining health percentage."""
    health_percent = health / max_health

    if health_percent > 0.5:
        bar_colour = "green"
    elif health_percent > 0.2:
        bar_colour = "orange"
    else:
        bar_colour = "red"

    turtle.initializeTurtle()

    bar_width = 300
    bar_height = 40

    # Draw grey background bar (empty health)
    turtle.fillcolor("grey")
    turtle.begin_fill()
    for _ in range(2):
        turtle.forward(bar_width)
        turtle.right(90)
        turtle.forward(bar_height)
        turtle.right(90)
    turtle.end_fill()

    # Draw coloured foreground bar (current health)
    filled_width = bar_width * health_percent
    turtle.fillcolor(bar_colour)
    turtle.begin_fill()
    for _ in range(2):
        turtle.forward(filled_width)
        turtle.right(90)
        turtle.forward(bar_height)
        turtle.right(90)
    turtle.end_fill()


# --- Preset values ---
level_up_threshold = 150
max_health = 100
health = 100
experience = 160
level = 1

# Test level up
level = level_up(experience, level, level_up_threshold)

# Test take damage
health = take_damage(health, 30)
print(f"After damage: {health} HP")

# Test apply poison
health = apply_poison(health, 10, 3)
print(f"After poison: {health} HP")

# Draw final health bar
draw_health_bar(health, max_health)
```

```python
# The final function takes a list and returns how many numbers in the list are even

def is_even(n):
    return n % 2 == 0

numbers = [3, 6, 8, 11]

def counter(numbers):
    count = 0

    for n in numbers:
        if is_even(n):
            count = count + 1

    return count

print("Even numbers:", counter(numbers))
```
