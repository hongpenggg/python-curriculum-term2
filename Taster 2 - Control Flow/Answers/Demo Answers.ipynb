# 🧠 If/Else: Giving the Computer a Brain

Right now, your code runs every single line from top to bottom. But in a real game, different things happen depending on the situation:
- If your health hits 0 → Game Over
- If you have a key → Open the chest
- Otherwise → Display "You need a key!"

This is called **conditional logic**, and it's the backbone of every game ever made.

## The `if` Statement

The simplest form: "If this condition is True, do this thing."

```python
if condition:
    # do something (this code is INDENTED)
```

⚠️ **The colon `:` and the indentation (4 spaces) are REQUIRED.** Python uses indentation to know what code belongs inside the `if`.

````python
health = 1

if health <= 0:
    print("Game Over!")
    print("Try again?")
print("Hi")
````

Try changing `health` to `50` and run it again. Nothing prints — because the condition is `False`, so Python skips the indented block.

## `if` / `else`

What if you want something to happen when the condition is **False** too?

```python
if condition:
    # do this if True
else:
    # do this if False
```

## `if` / `elif` / `else`

What if there are **more than two options**? Use `elif` (short for "else if") to check extra conditions.

```python
if condition_1:
    # do this
elif condition_2:
    # do this instead
elif condition_3:
    # or this
else:
    # if nothing above matched, do this
```

Python checks from **top to bottom** and runs ONLY the first block that matches.

````python
health = 30

if health <= 0:
    print("Game Over!")
elif health <= 25:
    print("Critical health! Find a potion!")
elif health <= 50:
    print("Health is low. Be careful!")
else:
    print("You're doing great!")
````

Try changing `health` to different values (`0`, `10`, `50`, `100`) and see which message prints each time!

## Combining It All

You can use **logical operators** inside your `if` conditions to check multiple things at once!

````python
level = 10
has_boss_key = True

if level >= 10 and has_boss_key:
    print("You may enter the Boss Room!")
elif level >= 10:
    print("You're strong enough, but you need the Boss Key!")
else:
    print("Keep leveling up, warrior!")
````

# Micro-challenge 3 🔍: The Gatekeeper (If/Else)

You approach the entrance to the Dragon's Lair. The Gatekeeper will only let you in under certain conditions.

1. Use `input()` to ask the player for their `level` (convert to `int`).
2. Use `input()` to ask if they have a `dragon_permit` (the user should type `"yes"` or `"no"`).

````python
level = int(input("Enter your level: "))
dragon_permit = input("Do you have a dragon permit? (yes/no) ").strip().lower()
````

Write an `if/elif/else` block with the following rules:
   - If the player's level is **20 or above** AND they have a dragon permit → Print `"The gates open. Enter if you dare!"`
   - Elif the player's level is **20 or above** but they do NOT have a permit → Print `"You're strong enough, but you need a Dragon Permit!"`
   - Elif the player's level is **below 20** but they have a permit → Print `"You have the permit, but you need to reach Level 20!"`
   - Else → Print `"You need both Level 20 and a Dragon Permit to enter."`


````python
if level >= 20 and dragon_permit == "yes":
    print("The gates open. Enter if you dare!")
elif level >= 20 and dragon_permit != "yes":
    print("You're strong enough, but you need a Dragon Permit!")
elif level < 20 and dragon_permit == "yes":
    print("You have the permit, but you need to reach Level 20!")
else:
    print("You need both Level 20 and a Dragon Permit to enter.")
````
