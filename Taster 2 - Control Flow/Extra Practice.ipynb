**11.(Validate triangles and compute perimeter)**  
Write a program that reads three edges for a triangle and determines whether the input is valid. The input is valid if the sum of any two edges is greater than the third edge. The program will compute the perimeter of the triangle if the input is valid. Otherwise, display that the input is invalid.
Sample output:

```
Enter side 1: 2
Enter side 2: 2
Enter side 3: 1
Perimeter = 5

Enter side 1: 1
Enter side 2: 2
Enter side 3: 1
Invalid triangle!

```

**12.(Determine grade)**  
Write a program that prompts the user to enter a score between 0 and 100 inclusive and output the corresponding grade as well as a Pass/Fail status. Output an error message if the entered score in not within the valid range. The grading system is as follows (S and U are not passes):

A: 70 - 100  
B: 60 - 69  
C: 55 - 59  
D: 50 - 54  
E: 45 - 49  
S: 35 - 44  
U: 0 - 34  

Sample output:

```
Enter score: 73
A
Pass

Enter score: 37
S
Fail

Enter Score: -1
Invalid! Score must be within 0 - 100.

```


**13.(BMI Calculator)**

**14.(Round function)**  
Python has an inbuilt function, round(), which returns a floating point number that is a rounded version of the specified number, with the specified number of decimals. However, it does not follow the usual convention.
For eg:
3.5, 5.5, 7.5 etc are rounded up to become 4, 6 and 8 respectively.
2.5, 4.5, 6.5 etc are rounded down to become 2, 4 and 6 respectively.

Write program code that will round off a number according to the usual convention.
For eg:
2.5, 3.5, 4.5, 5.5 etc are rounded off to 3,4,5 and 6 respectively.

Your code should also work for  
•	negative numbers  
•	rounding off to any decimal place
