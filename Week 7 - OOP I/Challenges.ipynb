```python
# Run these test cases to check your code! If you see an OK, you're good to go!

import unittest

class TestStudentClass(unittest.TestCase):

    def setUp(self):
        """Runs before EVERY test case to reset the global student registry."""
        Student.all_students = []

    def test_01_constructor_and_get_courses(self):
        """Test that a new student is initialized correctly with no courses."""
        alice = Student("Alice")
        self.assertEqual(alice.get_courses(), [])

    def test_02_add_courses_single_and_multiple(self):
        """Test adding single and multiple courses, ensuring no duplicates."""
        alice = Student("Alice")

        # Add a single course
        alice.add_courses("CS101")
        self.assertCountEqual(alice.get_courses(), ["CS101"])

        # Add multiple courses
        alice.add_courses("MATH101", "PHYS101")
        self.assertCountEqual(alice.get_courses(), ["CS101", "MATH101", "PHYS101"])

        # Add duplicate course (should be silently handled/ignored)
        alice.add_courses("CS101")
        self.assertCountEqual(alice.get_courses(), ["CS101", "MATH101", "PHYS101"])

    def test_03_drop_courses(self):
        """Test dropping courses, including courses the student isn't taking."""
        alice = Student("Alice")
        alice.add_courses("CS101", "MATH101", "PHYS101")

        # Drop a single course
        alice.drop_courses("MATH101")
        self.assertCountEqual(alice.get_courses(), ["CS101", "PHYS101"])

        # Drop a course not enrolled in (should silently ignore)
        alice.drop_courses("CHEM101")
        self.assertCountEqual(alice.get_courses(), ["CS101", "PHYS101"])

        # Drop multiple courses at once
        alice.drop_courses("CS101", "PHYS101")
        self.assertEqual(alice.get_courses(), [])

    def test_04_common_courses(self):
        """Test finding overlapping courses between two students."""
        alice = Student("Alice")
        bob = Student("Bob")

        # No overlap initially
        alice.add_courses("CS101", "MATH101")
        bob.add_courses("PHYS101", "CHEM101")
        self.assertEqual(alice.common_courses(bob), [])

        # Introduce overlap
        bob.add_courses("CS101")
        self.assertCountEqual(alice.common_courses(bob), ["CS101"])
        self.assertCountEqual(bob.common_courses(alice), ["CS101"])

    def test_05_is_coursemate(self):
        """Test if two students are correctly identified as coursemates."""
        alice = Student("Alice")
        bob = Student("Bob")

        self.assertFalse(alice.is_coursemate(bob))

        alice.add_courses("CS101")
        bob.add_courses("CS101")
        self.assertTrue(alice.is_coursemate(bob))

        # Drop the overlapping course
        bob.drop_courses("CS101")
        self.assertFalse(alice.is_coursemate(bob))

    def test_06_common_friends_basic_and_exclusion_rules(self):
        """
        Test the special rules for common_friends:
        1. A student cannot be their own friend.
        2. Alice and Bob cannot be returned in each other's common_friends list.
        """
        sara = Student("Sara")
        lora = Student("Lora")

        # Scenario from the assignment prompt: Sara and Lora only share one class.
        sara.add_courses("Bit by Bit Coding")
        lora.add_courses("Bit by Bit Coding")

        # They shouldn't return each other or themselves
        self.assertEqual(sara.common_friends(lora), [])
        self.assertEqual(lora.common_friends(sara), [])

    def test_07_common_friends_complex_network(self):
        """
        Test a more complex friendship graph.
        Graph Setup:
        - Alice and Bob share no classes.
        - Charlie shares 'CS101' with Alice and 'MATH101' with Bob (Charlie is a common friend).
        - David shares 'CS101' with Alice but nothing with Bob (David is NOT a common friend).
        """
        alice = Student("Alice")
        bob = Student("Bob")
        charlie = Student("Charlie")
        david = Student("David")

        alice.add_courses("CS101")
        bob.add_courses("MATH101")

        # Charlie bridges Alice and Bob
        charlie.add_courses("CS101", "MATH101")

        # David only connects with Alice
        david.add_courses("CS101")

        # Check common friends between Alice and Bob
        self.assertCountEqual(alice.common_friends(bob), ["Charlie"])
        self.assertCountEqual(bob.common_friends(alice), ["Charlie"])

    def test_08_dynamic_friendship_updates(self):
        """Test that dropping a course dynamically updates coursemate and friendship status."""
        alice = Student("Alice")
        bob = Student("Bob")
        charlie = Student("Charlie")

        alice.add_courses("CS101")
        bob.add_courses("MATH101")
        charlie.add_courses("CS101", "MATH101")

        # Charlie is currently a common friend
        self.assertCountEqual(alice.common_friends(bob), ["Charlie"])

        # Charlie drops CS101, severing the friendship with Alice
        charlie.drop_courses("CS101")

        # Charlie should no longer be a common friend
        self.assertEqual(alice.common_friends(bob), [])

unittest.main(argv=['first-arg-is-ignored'], exit=False)
```
