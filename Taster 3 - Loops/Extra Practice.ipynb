**3. Factorial**  
Write a program that asks the user to input a non-negative integer n and calculates the factorial of n.  

The definition of n factorial (n!) is as follows:

For a non-negative integer, $n! = n \times (n-1) \times (n-2) \times \dots \times 3 \times 2 \times 1$  
and  
$0! = 1$


**6. Sum of Digits**

Write a Python script that uses a `while` loop to find the sum of the digits of the number $3^{12}$ (which is $531,441$).

Your program should output 18.

*Could you do it without converting 531,441 to a string? Make use of the modulo and floor division operators!*
