**1.(Convert from kilograms to pounds)**  
Write a program that displays the following table (1 kilogram = 2.2 pounds).

Sample output:
```
Kilograms Pounds
1         2.2
2         4.4
3         6.6
...
9         19.8
10        22.0

```

````python
print(f"{'Kilograms':<10}{'Pounds'}")
for kg in range(1, 11):
    pounds = kg * 2.2
    print(f"{kg:<10}{round(pounds, 1)}") # alternatively, you could use str(kg).ljust(10, ' ')
````

**2.(Convert from miles to kilometres)**  
Write a program that displays the following two tables side-by-side (1 mile = 1.609 kilometres):

Sample output:
```
Miles    Kilometers    Kilometres    Miles
1        1.609         20            12.430
2        3.218         25            15.538

...
9        14.481        60            37.290
10       16.090        65            40.398

```

````python
print(f"{'Miles':<10}{'Kilometers':<15}{'Kilometres':<15}{'Miles'}")
for i in range(1, 11):
    miles = i
    km_from_miles = miles * 1.609
    km = 20 + (i - 1) * 5
    miles_from_km = km / 1.609
    print(f"{miles:<10}{km_from_miles:<15.3f}{km:<15}{miles_from_km:.3f}")
````

**3. Factorial**  
Write a program that asks the user to input a non-negative integer n and calculates the factorial of n.  

The definition of n factorial (n!) is as follows:

For a non-negative integer, $n! = n \times (n-1) \times (n-2) \times \dots \times 3 \times 2 \times 1$  
and  
$0! = 1$

















````python
n = int(input('Enter a non-negative number: '))

factorial = 1
for i in range(1, n + 1): # Writing the loop this way helps us handle n = 0 without error
    factorial *= i

print(f'{n}! = {factorial}')
````

**4. Number Pyramid**  
Write a program that asks the user to input a positive number n.
Print a pyramid of numbers with height n.
```
eg: n = 5
    1
   123
  12345
 1234567
123456789
```

````python
n = int(input('Enter a positive number: '))

for i in range(1, n + 1): # loops n times to get n rows
    print(' ' * (n - i), end='')
    for j in range(1, 2 * i): # loops 2i - 1 times to print the numbers in each row
        print(j, end='')
    print()
````

**5. Inverted Hollow Number Pyramid**

Write a program that asks the user to input a positive number n.
Print a inverted hollow pyramid pyramid of numbers with height n.

```
eg: n = 5
1       9
 1     7
  1   5
   1 3
    1
```


````python
n = int(input('Enter a positive number: '))

for i in range(n, 0, -1): # i = n, n-1, n-2,..., 1 (counts down)
    print(' ' * (n - i) + '1', end='')
    if i > 1:
        print(' ' * (2 * i - 3) + str(2 * i - 1))
````

**6. Sum of Digits**

Write a Python script that uses a `while` loop to find the sum of the digits of the number $3^{12}$ (which is $531,441$).

Your program should output 18.

*Could you do it without converting 531,441 to a string? Make use of the modulo and floor division operators!*

````python
num = 3 ** 12
result = 0
while num > 0:
    result += num % 10 # adds last digit to result
    num //= 10 # removes last digit

print(f'Sum of digits = {result}')
````
