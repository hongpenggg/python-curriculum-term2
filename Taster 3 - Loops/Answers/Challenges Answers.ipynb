# Collaborative Challenge 🚀: "The Boss Battle Simulator"

You are coding the combat engine for a boss fight! The boss has high HP, and the player must attack until the boss is defeated.

## 🛠️ The Mission Requirements
1. Create a variable `boss_hp` set to `100`.
2. Create a **while loop** that runs as long as `boss_hp > 0`.
3. Inside the loop:
   - Ask the player: `"Do you want to (1) Attack or (2) Run?"` (Use `input()`)
   - If they choose `"1"` (Attack), subtract `25` from `boss_hp` and print the remaining HP.
   - If they choose `"2"` (Run), print `"You fled the battle!"` and use `break` to exit the loop.
   - If they enter anything else, print `"Invalid choice!"` and use `continue` to ask again.
4. If the loop ends because `boss_hp` reached 0, print `"VICTORY! The boss has been defeated!"`.

```python
# write code here

boss_hp = 100

while boss_hp > 0:
    choice = input("Do you want to (1) Attack or (2) Run? ").strip()

    if choice == "1":
        boss_hp -= 25
        print(f"You attacked! Boss HP: {boss_hp}")
    elif choice == "2":
        print("You fled the battle!")
        break
    else:
        print("Invalid choice!")
        continue

if boss_hp <= 0:
    print("VICTORY! The boss has been defeated!")
```

### Fibonacci Sequence Generator

#### **Background**
The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, usually starting with 0 and 1.The mathematical definition is:$$F_n = F_{n-1} + F_{n-2}$$

So,
$$F(0)=0$$
$$F(1)=1$$
$$F(2)=F(1)+F(0)=1$$
$$F(3)=F(2)+F(1)=2$$

#### **Task**
Write a program that asks the user for a positive integer, $n$. Your program should then calculate and print the first $n$ numbers of the Fibonacci Sequence.

```python
# write code here

n = int(input('Enter a positive integer n: '))

fn_minus_2 = 0
fn_minus_1 = 1


if n >= 0:
    print(f'f(0) = {fn_minus_2}')
if n >= 1:
    print(f'f(1) = {fn_minus_1}')

for i in range(2, n + 1):
    fn = fn_minus_1 + fn_minus_2 # calculating current fib number
    fn_minus_2 = fn_minus_1
    fn_minus_1 = fn
    print(f'f({i}) = {fn}')

```
