Here's the format for a while loop:

    while <condition>:
        #The indented lines under the "while" statement executes every loop

A while loop consists of

**"while"** => The keyword starting the while loop.

**condition** => A boolean (statement that results in True or False) that signals whether the while loop should end.

**: (the colon)** => Syntax that separates the condition from the body. Goes right after the condition.

**body** => The code inside the while loop that will keep running while the condition is true.

**An example:**
Keep eating while there is still rice remaining

```
while rice > 0:
    rice = rice - 1
```
`rice = rice - 1` will keep repeating until `rice > 0` becomes `False`. That is, until the variable `rice` stores a number that is not more than 0 (i.e. 0 or negative numbers).

#For Loops
Jack grew up, and became more responsible. As a result, his parents allow him to go and play once he has finished **50** pieces of homework, instead of all of them! 😯😯😯

To help Jack do exactly **50** pieces of homework, we can use a **For Loop**.

Here's the format for a **For Loop**:

```
    for i in range(number) :  
        #The indented lines under the "for" statement executes every loop
        print("Hello World") #Prints "Hello World" every time the loop runs
```
A for loop consists of

**"for"** => The keyword that tells Python you are about to start a loop.


**"i"** => The iterator variable. In each "round" of the loop, this variable automatically updates to the current number in the sequence.

Note: Just like with arrays, you can name this anything (like for task in range(50):).

**range()** => A built-in function that creates a sequence of numbers. By default, range(50) starts at 0 and goes up to (but does not include) 50. This results in exactly 50 iterations.

**: (the colon)** => The required syntax that signals the end of the loop header and the beginning of the action.

**body** => The indented code block. Everything inside this "body" will repeat for every number in the range.

````python
print("Numbers from 1 to 4")


print("Numbers from 0 to 4")


print("Numbers from 0 to 9, incrementing by 2")
````

Notice that the **brackets** `()` behind `range` can take either 1,2, or 3 numbers.  

If it takes 1 number, that number will be the "end".  
`range(end)` will consist of all integer numbers from 0 to the integer before "end".  
*It is important to note that if you only provide "end" and not "start", the list of numbers wil lalways start from 0 and not 1.

If it takes 2 numbers, the first will be the "start", while the second will be the "end".
`range(start, end)` will consist of all integer numbers from "start" to the integer before "end".   

If it takes 3 numbers, they will be "start", "end", and "step".  
`range(start, end, step)` will consist of integer numbers from start to the integer before "end", incrementing by an amount equal to "step".

For loops using `range()` will repeat as many times as there are numbers given by the `range()` function.  

During each loop, the variable will become the number given by `range()` for that loop.  
For example, for the first loop of `for i in range(5):` , the variable "i" will be equal to 0.
