## What is a function?

A function is a named block of code that does one well-defined thing. Think of it like a tiny machine: you give it inputs (arguments), it does work, and it may return an output.

**Why use functions?**
- Reuse code instead of copying and pasting.
- Give a name to a meaningful operation.
- Make code easier to read and test.


### Simple example — `greet`

We define a function `greet` that takes a name and prints a friendly message. This demonstrates the basic `def` syntax.


```python
def greet(name):
  print(f'Hello {name}')

text = greet('Tom')
print(text)
```

### Returning values vs printing

Functions can either *perform an action* (like printing) or *compute and return a value*. Prefer returning values when the result will be used later, because it's more flexible and testable.


```python
def greet(name):
  return f'Hello {name}'

text = greet('Tom')
print(text)
```

### Variable Scope

Variables that you use as paramters cannot be accessed outside the function, however if you use *global* variables, you can use these variables throughout your code.

```python
global_name = "World"

def greet():
  local_name = "Alice" #exists only within this function
  print(f"Inside the function (local): Hello, {local_name}!")
  print(f"Inside the function (global): Hello, {global_name}!") #can access global variable within the function

def greet_with_parameter(name):
  print(f"Inside greet_with_parameter (local): Hello, {name}!") #parameter works as local variable within the function only

greet()
greet_with_parameter("Bob")

print(f"Outside the function: Hello, {global_name}!")
```

### Default arguments and keyword arguments

You can provide defaults for parameters, and call functions using keywords to improve clarity.


```python
def character(name, language = 'English'):
  print(f'Name: {name}, Speaks: {language}')

character('Ben', 'Spanish')
```

### `*args` and `**kwargs` — flexible parameters

- `*args` collects extra positional arguments as a tuple. A tuple is just a list that you can't modify.
- `**kwargs` collects extra keyword arguments as a dict.

Useful when you don't know how many arguments you'll get.


```python
def summarize(*args, **kwargs):
    print('positional args (tuple):', args)
    print('keyword args (dict):', kwargs)

summarize(1,2,3, name='Rio', role='student', age = 18)
```

Here are the Turtle functions you will be using (listed in random order!):

- `turtle.color(color)` - This function can either take a color (`'red'`) or a color code (`"#FF0000"`)

- `turtle.initializeTurtle()`
- `turtle.forward(size)` - This function takes in the number of pixels to move forward as an integer

```python
import ColabTurtlePlus.Turtle as turtle

def get_angle(sides):
  angle = 360/sides
  return angle

# Step 1: Define a function that accepts two inputs: 'size' and 'color'
def draw_shape(sides, size: int, color: str):
    # Initialise turtle
    turtle.initializeTurtle()
    # Use the 'color' parameter to set the turtle's ink/fill color
    turtle.color(color)
    # Tell the turtle to start tracking the area to be filled
    turtle.begin_fill()
    #Get angle of shape
    angle = get_angle(sides)
    # Create a loop that repeats exactly 4 times (one for each side)
    for i in range(sides):
        # Move the turtle forward by the 'size' amount
        turtle.forward(size)
        # Turn the turtle left by 90 degrees
        turtle.left(angle)
    # Tell the turtle to stop tracking and fill the shape with color
    turtle.end_fill()
# --- Test Area ---
# Call your function below with a size of 150 and the color "orange"
draw_shape(5, 150, 'red')
```

### Tips & common mistakes

- Don't use mutable default arguments like `def f(x, items=[]):` — it can lead to surprising shared-state bugs.
- Prefer `return` over `print` when a function computes a value you'll use later.
- Keep functions short and focused: a function should do one thing well.


#Recursive functions

Recursive functions are functions that call themselves to solve a smaller, simpler instance of the same problem. The recursive process stops once it reaches a base case.



#Example 1: Exponentiation
Here we are going to write a simple recursive function that takes 2 vales, `x` and `n` and returns x^n

```python
def exp(x, n):
    if n == 1:
        return x
    else:
        return x * exp(x, n - 1)

print(exp(2, 5))
```

#Example 2: Fibonacci sequence
Here we are going to write a recursive function that returns the *nth* term of the fibonacci sequence. Each term of the fibonacci sequence is given by the sum of the previous 2 terms, the first 2 terms of the sequence are 1 and 1. Hence the sequence goes 1,1,2,3,5,8....

```python
def fib(n):
    if n <= 1:
        return n
    else:
        return fib(n - 1) + fib(n - 2)

print(fib(10))
```

1b. Create a function that takes a nested list and returns a flattened list.  
Eg: `[1, [2, 3], [4, [5, 6]]]` -> `[1,2,3,4,5,6]`

```python
# Method 1
def flatten(lst):
  if lst == []:
    return lst

  if type(lst[0]) == list:
    return flatten(lst[0])+flatten(lst[1:])

  return [lst[0]] + flatten(lst[1:])

flatten([1,[2,3],[4,[5,6]]])
```

```python
# Method 2
def flatten(lst):
    result = []
    if type(lst) == list:
        for elem in lst:
            result.extend(flatten(elem))
    else:
        result.append(lst)
    return result
flatten([1,[2,3],[4,[5,6]]])
```
