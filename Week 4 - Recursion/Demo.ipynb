## What is Recursion?

A **recursive function** is a function that calls itself. Every recursive solution needs two things:

1. **Base case** — the simplest version of the problem that we can answer directly (this stops the recursion).
2. **Recursive case** — the function calls itself on a *smaller* input, moving toward the base case.

> ⚠️ If you forget the base case, the function calls itself forever → `RecursionError` (a stack overflow)!

## Example 1: Countdown

Let's print numbers from `n` down to `1`, then `Blastoff!`. The base case is `n == 0`.

````python
def countdown(n):
    if n == 0:            # base case
        print("Blastoff!")
        return
    print(n)             # do something
    countdown(n - 1)     # recursive case (smaller input)

countdown(5)
````

## Example 2: Factorial

`n!` (n factorial) is `n * (n-1) * ... * 1`. Notice that `n! = n * (n-1)!`, and `0! = 1`. That is the perfect shape for recursion.

````python
def factorial(n):
    if n == 0:                       # base case
        return 1
    return n * factorial(n - 1)      # recursive case

for i in range(7):
    print(i, '->', factorial(i))
````

## Example 3: Fibonacci

The Fibonacci sequence is `1, 1, 2, 3, 5, 8, ...` where each term is the sum of the previous two. The first two terms are both `1` (base cases).

````python
def fib(n):
    if n <= 2:                 # base cases: fib(1) = fib(2) = 1
        return 1
    return fib(n - 1) + fib(n - 2)   # recursive case

for i in range(1, 8):
    print(fib(i), end=' ')
````

## Tracing the Call Stack

When `factorial(3)` runs, Python builds up a *stack* of unfinished calls:

```
factorial(3)  ->  3 * factorial(2)
factorial(2)  ->  2 * factorial(1)
factorial(1)  ->  1 * factorial(0)
factorial(0)  ->  1   (base case!)
```
Then it unwinds: `1 -> 1 -> 2 -> 6`. Drawing the call stack on paper is the #1 way to understand recursion.

## ⚠️ Common Mistakes

- Forgetting the base case → infinite recursion.
- The recursive case not moving *toward* the base case (e.g. `factorial(n)` calling `factorial(n)` instead of `factorial(n-1)`).
- Returning `None` because you forgot `return` in the recursive case.

# Micro-challenge 3 🔍: Flatten a Nested List

Write a recursive function `flatten(lst)` that turns a nested list into a flat one.

Example: `flatten([1, [2, 3], [4, [5, 6]]])` returns `[1, 2, 3, 4, 5, 6]`.

*Hint:* use `isinstance(item, list)` to check whether to recurse.

````python
# write code here
````

````python
# @title Hint if you are really stuck
def flatten(lst):
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))   # recurse
        else:
            result.append(item)            # base case element
    return result
````
