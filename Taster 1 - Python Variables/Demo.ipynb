# 🧠 Giving the Computer a Memory

Great job making the computer talk! But right now, it has the memory of a goldfish—it forgets everything immediately.

To fix this, we use **Variables**.

## What is a Variable?

Imagine a **box**.
1. You stick a **Label** on the outside (e.g., `favorite_food`).
2. You put **Data** inside the box (e.g., `"Sushi"`).

Whenever you ask the computer for `favorite_food`, it looks in the box and gives you `"Sushi"`.

The `=` sign means **"put this value into this box."** It does NOT mean "equals" in the math sense!

You can also **overwrite** a variable. Whatever you put in the box last is what comes out!

### ⚠️ Important: No Spaces!
Variable names **cannot have spaces**.
- ✅ `favorite_food` (Good! We use underscores `_`)
- ❌ `favorite food` (Bad! The computer gets confused)

*Pro Tip: Variable names should be descriptive. `x` is fine for math, but `player_name` is way more readable in a game!*

# 👂 Teaching the Computer to Listen

Your computer now has a **Mouth** (`print`) and a **Memory** (variables). Now let's give it **Ears**.

Up until now, *you* (the coder) have been typing all the data. Now we want to let the *user* (the person playing your game) answer questions.

We use a command called `input()`.

**How it works:**

`input()` **always** returns a **string**, even if the user types a number!

If you want to do math with the user's answer, you need to **convert** it using `int()` or `float()`.

# 🧠 If/Else: Giving the Computer a Brain

Right now, your code runs every single line from top to bottom. But in a real game, different things happen depending on the situation:
- If your health hits 0 → Game Over
- If you have a key → Open the chest
- Otherwise → Display "You need a key!"

This is called **conditional logic**, and it's the backbone of every game ever made.

## The `if` Statement

The simplest form: "If this condition is True, do this thing."

```python
if condition:
    # do something (this code is INDENTED)
```

⚠️ **The colon `:` and the indentation (4 spaces) are REQUIRED.** Python uses indentation to know what code belongs inside the `if`.

````python
health = 0

if health <= 0:
    print("Game Over!")
````

Try changing `health` to `50` and run it again. Nothing prints — because the condition is `False`, so Python skips the indented block.

## `if` / `else`

What if you want something to happen when the condition is **False** too?

```python
if condition:
    # do this if True
else:
    # do this if False
```

````python
# Try it out!
````

## `if` / `elif` / `else`

What if there are **more than two options**? Use `elif` (short for "else if") to check extra conditions.

```python
if condition_1:
    # do this
elif condition_2:
    # do this instead
elif condition_3:
    # or this
else:
    # if nothing above matched, do this
```

Python checks from **top to bottom** and runs ONLY the first block that matches.

````python
health = 30

if health <= 0:
    print("Game Over!")
elif health <= 25:
    print("Critical health! Find a potion!")
elif health <= 50:
    print("Health is low. Be careful!")
else:
    print("You're doing great!")
````

Try changing `health` to different values (`0`, `10`, `50`, `100`) and see which message prints each time!

## Combining It All

You can use **logical operators** inside your `if` conditions to check multiple things at once!

````python
level = 10
has_boss_key = True

if level >= 10 and has_boss_key:
    print("You may enter the Boss Room!")
elif level >= 10:
    print("You're strong enough, but you need the Boss Key!")
else:
    print("Keep leveling up, warrior!")
````
