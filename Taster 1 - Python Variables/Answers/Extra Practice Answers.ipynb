**3. (Converting Fahrenheit to Celsius)**

Write a program that reads a Fahrenheit degree as a float from standard input, then converts it to Celsius (round to one decimal place) and displays the result in standard output. The formula for the conversion is:

celsius = (5/9) * (fahrenheit - 32)

````python
fahrenheit = float(input("Enter temperature in Fahrenheit: "))
celsius = (5 / 9) * (fahrenheit - 32)
print(f"{fahrenheit}°F = {round(celsius, 1)}°C")
````

**4.	(Computing the area, perimeter and volume)**

a)	Write a program that reads in the width and height of a rectangle. Calculate and print out its perimeter and area.

```
Enter width: 2.4
Enter height: 3.7
Rectangle(2.4, 3.7) has perimeter of 12.2 and area of 8.88

```

````python
width = float(input("Enter width: "))
height = float(input("Enter height: "))

perimeter = 2 * (width + height)
area = width * height

print(f"Rectangle({width}, {height}) has perimeter of {round(perimeter, 2)} and area of {round(area, 2)}")
````

b)	Write a program that reads in the radius and length of a cylinder and computes its volume using the formulae:

area = radius * radius * pi  
volume = area * length

Define a constant to store pi for calculation


```
Enter radius of cylinder: 5
Enter length of cylinder: 5
Volume of cylinder = 392.5

```

````python
from math import pi # You can actually import pi directly from the math library

PI = pi # Constants should be in all caps

radius = float(input("Enter radius of cylinder: "))
length = float(input("Enter length of cylinder: "))

area = radius * radius * PI
volume = area * length

print(f"Volume of cylinder = {round(volume, 1)}")
````

**5.	(Converting miles into kilometres)**

Write a program that reads a number in miles, converts it to kilometres, and outputs the result. One mile is 1.60934 kilometres. Display your answer correct to 3 decimal places.

````python
MILES_TO_KM = 1.60934

miles = float(input("Enter distance in miles: "))
kilometres = miles * MILES_TO_KM

print(f"{miles} miles = {kilometres:.3f} kilometres")
````

**6.	(Converting an uppercase letter to lowercase)**

Write a program that converts an uppercase letter from standard input to a lowercase letter by making use of its ASCII value.

http://www.asciitable.com

````python
upper = input("Enter an uppercase letter: ").strip()

lower = chr(ord(upper) + 32)

print(f"{upper} -> {lower}")
````

**10.	Inbuilt String methods**

String class has inbuilt search methods (count(), find() and rfind()). Each of these methods return objects of different types. Use these methods for the following exercises.  
a) 	Replace ‘good’ with ‘great’ in the string ‘Have a good day!’   
b) 	Given the string ‘The fox jumps over the lazy dog. The lazy dog barked at the fox.’  
(i) Find the index of the first occurrence of the letter ‘z’ in the string.   
(ii) Find the index of the last occurrence of the letter ‘o’ in the string.   
(iii) Find the number of occurrences to the word ‘dog’  


````python
# a) Replace 'good' with 'great'
a = 'Have a good day!'
print(a.replace('good', 'great'))

# b) String operations
b = 'The fox jumps over the lazy dog. The lazy dog barked at the fox.'

# (i) First occurrence of 'z'
print(b.find('z'))

# (ii) Last occurrence of 'o'
print(b.rfind('o'))

# (iii) Number of occurrences of 'dog'
print(b.count('dog'))
````
