# 🧠 Giving the Computer a Memory

Great job making the computer talk! But right now, it has the memory of a goldfish—it forgets everything immediately.

To fix this, we use **Variables**.

## What is a Variable?

Imagine a **box**.
1. You stick a **Label** on the outside (e.g., `favorite_food`).
2. You put **Data** inside the box (e.g., `"Sushi"`).

Whenever you ask the computer for `favorite_food`, it looks in the box and gives you `"Sushi"`.

The `=` sign means **"put this value into this box."** It does NOT mean "equals" in the math sense!

````python
favourite_food = 'Sushi'
print(favourite_food)
````

You can also **overwrite** a variable. Whatever you put in the box last is what comes out!

````python
favourite_food = 'Pizza'
print(favourite_food)
````

### ⚠️ Important: No Spaces!
Variable names **cannot have spaces**.
- ✅ `favorite_food` (Good! We use underscores `_`)
- ❌ `favorite food` (Bad! The computer gets confused)

*Pro Tip: Variable names should be descriptive. `x` is fine for math, but `player_name` is way more readable in a game!*

### ⚠️ Strings vs Numbers
This is a **super common mistake**. Watch what happens when you "add" two strings vs two numbers:

````python
print(10 + 5)
print(type('10' + '5'))
````

The first one gives `15` (math addition). The second gives `"105"` (it glues the text together—this is called **concatenation**).

*Pro Tip: If you ever see weird behaviour with numbers, check if they're accidentally strings!*

# 👂 Teaching the Computer to Listen

Your computer now has a **Mouth** (`print`) and a **Memory** (variables). Now let's give it **Ears**.

Up until now, *you* (the coder) have been typing all the data. Now we want to let the *user* (the person playing your game) answer questions.

We use a command called `input()`.

**How it works:**
1. The computer prints your question.
2. The program **PAUSES** and waits.
3. The user types an answer and hits `Enter`.
4. The computer saves that answer into a variable.

````python
name = input('What is you name: ')
# print('Hi,', name + '!')
print(f"The lower case of your name is {name.lower()}")
````

### ⚠️ The `input()` Trap

`input()` **always** returns a **string**, even if the user types a number!

If you want to do math with the user's answer, you need to **convert** it using `int()` or `float()`.

````python
number = int(input('Enter a number: '))
# print(type(number))
print(f'{number} + 5 = {number + 5}')
````

*Pro Tip: You can combine `input()` and `int()` into one line:*

````python
# Imagine a player with 100 health takes 35 damage
health = 100
damage = 35
health = health - damage
print("Health remaining:", health)
````

````python
# Splitting 100 gold coins among 3 players
total_gold = 100
players = 3

each_gets = total_gold // players   # whole coins per player
leftover = total_gold % players   # coins left over

print("Each player gets:", each_gets, "gold")
print("Leftover:", leftover, "gold")
````

*Pro Tip: You can use the shortcut `+=`, `-=`, `*=` etc. to update a variable in place:*
```python
health -= damage   # same as health = health - damage
score += 10        # same as score = score + 10
```
Try it below!

````python
product = 3

product *= 4

print(product)

sum = 5

sum += 5

print(sum)

answer = 70

answer -= 3

print(answer)
````

# Micro-challenge 2 🔍: The Potion Mixer (Input & Operators)
You run a potion shop and need a calculator to figure out prices.

1. Use `input()` to ask the user: `"How many potions do you want to buy? "`. Store it in a variable called `quantity`.

*Remember: `input()` returns a string! You'll need to convert it.*

````python
# write code here
quantity = input('How many potions do you want to buy? ')

quantity = int(quantity)

print(type(quantity))
````

2. Each potion costs `15` gold. Calculate the `total_cost` and print it.

For example: `"Total cost: 45 gold"`

````python
# write code here

total_cost = quantity * 15

print(f"Total cost: {total_cost} gold")
````

3. The shop has a "Buy 3 get 1 free" deal. Use the `//` (floor division) and `%` (modulo) operators to calculate:
   - How many **free** potions does the customer get?
   - How many potions do they actually **pay for**?
   - What is the **discounted total cost**?

*Hint: If the customer buys 10 potions, they get 10 // 4 = 2 free ones, and pay for 10 - 2 = 8.*

````python
# write code here

free_potions = quantity // 4
print(free_potions)

paid_potions = quantity - free_potions
print(paid_potions)

discounted_total_cost = paid_potions * 15
print(discounted_total_cost)
````

**Bonus Task**: Calculate and print how much gold the customer **saved** compared to the original total cost.

````python
# write code here

gold_saved = total_cost - discounted_total_cost

print(f"You saved {gold_saved} gold!")
````
