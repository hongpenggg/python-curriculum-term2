**Objective**: Create a Python program that can "scramble" and "unscramble" secret messages using the **ASCII** values of letters.

**The Background**
In computers, letters aren't stored as "A" or "B." They are stored as numbers (Data Representation!). For example, in the ASCII standard, the letter *'A'* is represented by the number 65.

A **Caesar Cipher** works by taking a letter, finding its number, adding a "Secret Shift" to that number, and then turning it back into a new letter.

**Example with a Shift of 3:**
- 'A' (65) + 3 = 68.
- 68 is the letter **'D'**.
- So, "CAT" becomes "FDW".

**Your Tasks**
1. **The Input**: Ask the user for a message (string) and a shift amount (integer).
2. **The Loop**: Use a for loop to look at every character in the message one by one.
3. **The Magic Functions**: * Use ord(char) to turn a letter into its ASCII number.
    - Add the shift amount to that number.
    - Use chr(number) to turn the new number back into a character.

4. **The Output**: Print the final "encrypted" message to the screen.
