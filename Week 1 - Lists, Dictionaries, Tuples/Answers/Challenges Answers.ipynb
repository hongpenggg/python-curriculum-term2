# Collaborative Challenge 🚀: "The Ultimate Loot Tracker"

Your team is building the inventory system for a new multiplayer RPG. You need to create a system that lets a player view their inventory, buy new items, and sell items to a shop. Run the cell below to get started!

```python
# Initial character data
player_stats = {
    "username": "Hero_123",
    "gold": 70,
    "inventory": []
}

# The shop's price list
item_prices = {
    "Shield": 50,
    "Bread": 5,
    "Rusty Sword": 20,
    "Health Potion": 30
}
```

## 🛠️ The Mission Requirements

Your code must perform the following tasks in a loop so the player can keep performing actions until they type `"Quit"`:
1. **Display Status**: Use a loop to print every item in the inventory list and show the current gold balance.
2. **User Input**: Ask the player what they want to do: `"Buy"`, `"Sell"`, or `"Quit"`.
3. **The "Buy" Feature**: Ask the user for the name of the item they want to buy. If the player:
  - Has enough gold to buy the item
  - Does not already own the item (item is not in `inventory`)
  - The item exists in `item_prices`

Then, add the item to the players `inventory` and decrease the player's `gold` by the price of the item.

*Hint: Find a way to check if the item exists in the item_prices dictionary before adding it*

4. **The "Sell" Feature**: If they have it in their `inventory`, remove it from the list. Find the price of that item in `item_prices` and add that amount to the `gold` value in the `player_stats` dictionary.

*Hint: How do you delete a specific item from a list by its name?*

6. **Exit Strategy**: If they type `"Quit"`, break the loop and print a "Game Over" message.

*Note: The player initially starts off with an empty inventory*

```python
# It is good practice to Google whatever you are unfamiliar with.

while True:
    print("\n--- Current Status ---")
    print(f"Gold: {player_stats['gold']}")
    print(f"Inventory: {player_stats['inventory']}")
    print("----------------------")

    action = input("What do you want to do (Buy, Sell, or Quit): ")

    if action == "Buy":
        buy_item = input(f"Pick an item from the shop: {item_prices}: ")
        if buy_item in item_prices:
            if buy_item not in player_stats["inventory"]:
                item_price = item_prices[buy_item]
                if player_stats["gold"] >= item_price:
                    player_stats["gold"] -= item_price
                    player_stats["inventory"].append(buy_item)
                    print(f"You bought {buy_item} for {item_price} gold.")
                else:
                    print(f"You don't have enough gold to buy {buy_item}. You need {item_price} gold but only have {player_stats['gold']}.")
            else:
                print(f"You already own {buy_item}.")
        else:
            print(f"{buy_item} is not available in the shop.")

    elif action == "Sell":
        sell_item = input(f"Which item do you want to sell? Your inventory: {player_stats['inventory']}: ")
        if sell_item in player_stats["inventory"]:
            if sell_item in item_prices: # Ensure it's a sellable item that exists in the shop prices
                item_price = item_prices[sell_item]
                player_stats["gold"] += item_price
                player_stats["inventory"].remove(sell_item)
                print(f"You sold {sell_item} for {item_price} gold.")
            else:
                print(f"You can't sell {sell_item}. It's not a recognized shop item.")
        else:
            print(f"You don't have {sell_item} in your inventory to sell.")

    elif action == "Quit":
        print("Game Over! Thanks for playing.")
        break

    else:
        print("Invalid action. Please choose 'Buy', 'Sell', or 'Quit'.")
```

### Fibonacci Sequence Generator *Improved*

#### **Background**
The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, usually starting with 0 and 1.The mathematical definition is:$$F_n = F_{n-1} + F_{n-2}$$

So,
$$F(0)=0$$
$$F(1)=1$$
$$F(2)=F(1)+F(0)=1$$
$$F(3)=F(2)+F(1)=2$$

#### **Task**
Remember the Fibonacci Sequence Generator we made last week? Lets optimise it!

1. We can simplify our loop! Instead of keeping track of two separate variables, we can pull the values we need **directly** from the list we're building!

2. Currently, we're just printing each number individually, which is a bit inefficient. This time, let's modify the logic to store the sequence in a **list** first, then output the entire result with a **single** print statement.



```python
# Generate Fibonacci sequence up to n terms directly
num_terms = 10
fib_sequence = []

if num_terms >= 1:
    fib_sequence.append(0)
if num_terms >= 2:
    fib_sequence.append(1)

for i in range(2, num_terms):
    # Pull values directly from the list
    next_fib = fib_sequence[i-1] + fib_sequence[i-2]
    fib_sequence.append(next_fib)

print(f"Fibonacci sequence up to {num_terms} terms: {fib_sequence}")
```
