# Lists
Think of a **Python list** as your digital inventory. Instead of having a bunch of loose items (variables) floating around, you put them all into one organised container.

## What is a List?

- **It’s a Container**: It holds multiple items under one name.
- **Mixed Bag**: You can throw strings, numbers, or even other lists inside one list.
- **Ordered**: Python remembers exactly what order you put things in.

In Python, lists are defined by **square brackets** `[]`, and items are separated by commas.

## Slicing: Select Your Loadout

Slicing lets you grab a specific "sub-list" (a piece of the original).

The syntax is: `list[start : stop: step]`
- The start index is included.
- The stop index is excluded (Python stops right before this index).
- The step size, specifying the interval between elements.

*Try leaving the start, stop and step empty to see what happens!*

# Dictionaries

If a Python list is an ordered junk drawer, a Python dictionary is a labeled locker system.

In a list, you find things by their position (0, 1, 2). In a dictionary, you find things by a specific **Key** (like a name tag) to get a **Value** (the actual data).  

Dictionaries use curly braces `{}`. Each entry is a pair: a **key**, a **colon**, and a **value**.

```python
# A dictionary representing a game character's stats
character = {
    "name": "Zelda",
    "level": 50,
    "item": "Master Sword",
    "is_alive": True
}
```

Key features:
- **Key-Value Pairs**: Every piece of data has a unique label.
- **Unordered (Technically)**: While modern Python remembers the order you typed them, you don't use numbers to find things; you use the labels.
- **Unique Keys**: You can't have two "name" tags in the same locker—Python wouldn't know which one to open!

Imagine you are building an RPG. A list of items is fine, but a dictionary lets you store **details** about those items.

# Lists vs. Dictionaries: Which to use?
- Use a **List** if you have a collection of similar items where the **order** matters (like a leaderboard or a queue of players).
- Use a **Dictionary** if you have a piece of data that needs a **description** (like a user profile, a contact card, or a game character's attributes).
