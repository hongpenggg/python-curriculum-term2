# Let's learn how to analyse `data/student_grades.csv`!

First, we use the `csv` library to read and save the data into a list.

```python
import csv

student_grades =[]

# If you get a FileNotFoundError
# Download student_grades.csv from the data folder in the drive
# Drag and drop the file into the 'Files' tab on the left side of the notebook
# Run the code again
with open("student_grades.csv") as file:
    reader = csv.reader(file)
    headers = next(reader) # Remove header for easier data manipulation
    for i in reader:
        student_grades.append(i)

print(student_grades)

data =[
    ["Name", "Age"],
    ["Kevin", "Old"]
]

data2 = ["JFlkdjsd","nfhdsjkhsfdjka"]

with open("info.csv", "w") as file:
    writer = csv.writer(file)
    # all at once
    writer.writerows(data)
    # one by one
    for row in data:
        writer.writerow(row)
```

Notice how the scores are saved as a string instead of an integer. Let's use `map()` to convert all the appropriate stings to integers (in this case only `Percentage_Score`)

```python
# lambda is an anonymous function
student_grades = list(map(lambda record: [int(elem) if elem.isdigit() else elem for elem in record], student_grades))
print(student_grades)
```

We can use `filter()` to find out which students passed.

```python
students_passed = list(filter(lambda record: record[3] >= 50, student_grades))
print(students_passed)
```

We can find the mean, median and modal percentage score of all students

```python
from statistics import mean, median, mode

scores = list(map(lambda record: record[3], student_grades))
print(f'Mean: {mean(scores)}')
print(f'Median: {median(scores)}')
print(f'Mode: {mode(scores)}')
```

## Micro-challenge
We can also find which of the 5 subjects has the highest mean score and which has the lowest. Ensure that you find the 5 subjects dynamically, don't hard code it!

*Hint: Use a dictionary where the keys are the Subject names and the values are lists of scores.*

```python
# Find all the unique subjects
subjects = set(map(lambda record: record[2], student_grades))

scores_by_subject = {}
for subj in subjects:
    # Set each subject to an empty list
    scores_by_subject[subj] = []

for record in student_grades:
    # Append each score to the appropriate subject list
    scores_by_subject[record[2]].append(record[3])

print(scores_by_subject)

highest_mean_subj_name = ''
highest_mean_subj_score = 0
lowest_mean_subj_name = ''
lowest_mean_subj_score = 100
for subject, scores in scores_by_subject.items():
    mean_score = mean(scores)

    # Update highest score
    if mean_score >= highest_mean_subj_score:
        highest_mean_subj_score = mean_score
        highest_mean_subj_name = subject

    # Update lowest score
    if mean_score < lowest_mean_subj_score:
        lowest_mean_subj_score = mean_score
        lowest_mean_subj_name = subject

print(f'{highest_mean_subj_name} has the highest mean score of {highest_mean_subj_score}')
print(f'{lowest_mean_subj_name} has the lowest mean score of {lowest_mean_subj_score}')
```
