# Let's learn how to analyse a text file!

## An important function for dealing with files in Python is `open()`.

### There are 4 primary ways to interact with files:
- "w", which is for writing/overwriting
- "r", which is for reading
- "a", which is for appending content
- "x", which is for creating files

### Moreover, the 2 types of files you may deal with are
- "t", which are text files
- "b", which are binary files such as images

# Section 1 : Opening Files
This is how one can use `with open()`. The file value will be returned so ensure you store it in a variable as shown below.

Write mode creates the file if it does not already exist, and OVERWRITES the file if it does already exist.

Therefore, be extremely cautious when using it instead of append or create.

Remember to close your file after opening it and using it. This frees memory and prevents corruption.

*Note: After a `with open()` block, the file automatically closes for you.*

```python
# Using write to create and open a file

# 'w' means that you are writing to the file
# If the file does not exist, it will be created
with open('test.txt', 'w') as f:
    f.write('Hello world!\n') # You need to add the newline character yourself
    f.write('Python is great.\n')
    f.write('File reading is easy.')
```

# Section 2 : Basic Reading

This program uses `read()`, which allows the contents of the file to be stored into the variable as shown below.

```python
with open('test.txt', 'r') as f:
    print(f.read())
```

# Section 3 : Advanced Reading
Below shows 3 alternative ways to read, allowing you to specify what part of the file you want read, and how the data returned should be formatted. This can help a lot in data cleaning.

<table>
  <thead>
    <tr>
      <th>Method</th>
      <th>Action</th>
      <th>Return Type</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>f.read()</code></td>
      <td>Reads the entire file content into memory.</td>
      <td>Single String</td>
    </tr>
    <tr>
      <td><code>f.readline()</code></td>
      <td>Reads exactly one line at a time.</td>
      <td>Single String</td>
    </tr>
    <tr>
      <td><code>f.readlines()</code></td>
      <td>Reads all lines and returns them as a list of strings.</td>
      <td>List of Strings</td>
    </tr>
  </tbody>
</table>

*Note: `readline()` and `readlines()` will both return any trailing `"\n"` at the end of each line.*

```python
# f.read()
with open("test.txt") as f:
    content = f.read()

print(content)
print(type(content))
print()

# f.readline()
with open("test.txt") as f:
    line1 = f.readline()
    line2 = f.readline()
    line3 = f.readline()

print(line1)
print(line2)
print(line3)
print()

# f.readlines()
with open("test.txt") as f:
    lines = f.readlines()

print(lines)
print(type(lines))

print(lines[0])
print(lines[-1])
```

# Section 4 : Advanced Writing
Below shows the 2 main ways to write to a file.

<table>
  <thead>
    <tr>
      <th>Method</th>
      <th>Action</th>
      <th>Expected Input Type</th>
    </tr>
  </thead>
  <tbody>
    <tr>
      <td><code>f.write()</code></td>
      <td>Writes a specific string to the file.</td>
      <td>Single String</td>
    </tr>
    <tr>
      <td><code>f.writelines()</code></td>
      <td>Writes a collection of strings to the file sequentially.</td>
      <td>List (or Iterable) of Strings</td>
    </tr>
  </tbody>
</table>

*Note: For `writelines()`, Python won't help you add `"\n"` to the end of the each string in the list, you have to add it yourself.*

```python
# f.write() was demonstrated above

# f.writelines()
lines = [
    "Hello, World!\n",
    "Python is great.\n",
    "File writing is easy."
]

with open("writing_test.txt", "w") as f:
    f.writelines(lines)

# Let's see what's in the file!
with open('writing_test.txt') as f:
    print(f.read())
```

# Section 5 : Appending
Appending is similar to writing, as it is an input method. However, as it only adds to the end of a file, it is considered safer to use.

This still uses `write()`, but the type in `open()` is `"a"`.

```python
with open('writing_test.txt', 'a') as f:
    f.write('\nThis is appended text')

# Let's see what's in the file!
with open('writing_test.txt') as f:
    print(f.read())
```
