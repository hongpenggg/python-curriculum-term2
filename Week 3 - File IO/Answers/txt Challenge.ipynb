```python
# If you get a FileNotFoundError
# Download words.txt from the data folder in the drive
# Drag and drop the file into the 'Files' tab on the left side of the notebook
# Run the code again
with open("words.txt", "r") as file:
    lines = file.readlines()

# Remove trailing \n
lines = list(map(lambda word: word.strip(), lines))
print(lines)
```

Using code, find the duplicated words and remove the duplicates so that each word only appears once.
```python
# loop thru all the words
# add it to a new list
# check if its in the list
# sort the list

unique = []

for word in lines:
    if word not in unique:
        unique.append(word)

unique.sort()
print(unique)
print(len(unique))

with open('words_cleaned.txt', 'w') as f:
    f.write('\n'.join(unique))
```

## Some ideas for data analysis:
- Find the total number of words

```python
with open('words_cleaned.txt') as f:
    words = f.read().split('\n')

print(f'Total number of words: {len(words)}')
```

- Find all words that start with a specified letter

```python
letter = 'a'
startswith_letter = 0


for word in words:
    if word.startswith(letter):
        startswith_letter += 1

print(f'Number of words that starts with {letter}: {startswith_letter}')
```

- Find the shortest, longest words and their lengths & find the mean, median, modal length of each word

```python
from statistics import mean, median, mode, multimode

def analyze_words(words):
    if not words:
        print("No words provided.")
        return

    lengths = [len(w) for w in words]

    shortest = min(words, key=len)
    longest  = max(words, key=len)

    avg_len    = mean(lengths)
    median_len = median(lengths)
    modes      = multimode(lengths)   # returns all modes if there's a tie

    print("=" * 40)
    print("       WORD LENGTH ANALYSIS")
    print("=" * 40)

    print(f"\n{'Shortest word:':<20} '{shortest}'  (length {len(shortest)})")
    print(f"{'Longest word:':<20} '{longest}'  (length {len(longest)})")

    print(f"\n{'Mean length:':<20} {avg_len:.2f}")
    print(f"{'Median length:':<20} {median_len}")

    if len(modes) == 1:
        print(f"{'Modal length:':<20} {modes[0]}")
    else:
        print(f"{'Modal lengths (tie):':<20} {modes}")

    print("=" * 40)

analyze_words(words)
```

- Strip all vowels from each word and write these “consonant-only” words into a new file -> `words_no_vowels.txt`


```python
vowels = 'aeiou'

to_write = ''
for char in '\n'.join(words):
    if char not in vowels:
        to_write += char

with open('words_no_vowels.txt', 'w') as f:
    f.write(to_write)
```
