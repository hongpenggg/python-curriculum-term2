# Sorting Algorithms

In this notebook, we will explore all about certain functions that we can use to sort a list of items. This is a very important concept -- there are many use-cases of sorting algorithms in real life.

There are many sorting algorithms, but in general, we will cover just some of them:

* **Bubble Sort:** Think of this as the "heavy items sink, light items float" method. It repeatedly steps through the list, compares items side-by-side, and swaps them if they are in the wrong order. The largest numbers slowly "bubble" up to the end of the list.
    
    * **Use Case:** It is notoriously slow for big lists, so professional programmers rarely use it in real apps. However, it is a fantastic, simple tool for learning *how* algorithms work.

Bubble Sort visualisation: https://youtube.com/shorts/KmoZaD-PcCg?si=5x__yGbv0z0fAXrM

* **Insertion Sort:** Imagine you are sorting a hand of playing cards. You pick up one card at a time and insert it into its correct spot among the cards you are already holding. That is exactly what Insertion Sort does!
    * **Use Case:** Because it is very efficient when a list is already *mostly* sorted, it's great for adding a few new high scores to an existing leaderboard in a video game.

* **Merge Sort:** This uses a clever strategy called "Divide and Conquer." (reminder: we've seen this before somewhere!) Instead of trying to sort the whole list at once, it chops the list in half over and over until every item is by itself. Then, it stitches (merges) them back together in the correct order.
    
    * **Use Case:** Merge sort is incredibly reliable and handles massive amounts of data well. It is often used behind the scenes in big e-commerce databases (like sorting thousands of Amazon products by price).

* **Quick Sort:** Another "Divide and Conquer" heavyweight! It picks one item to act as a "pivot." It throws all the smaller numbers to the left of the pivot, and all the bigger numbers to the right. Then it repeats the process for the left and right sides.
    * **Use Case:** As the name implies, it is usually extremely fast. It is frequently used in commercial software for general-purpose sorting, like organizing your computer's files by date or size.

* **Timsort (Python's Secret Weapon):** You might be wondering, "When I just type `.sort()` in Python, which algorithm does it use?" It uses Timsort! Timsort is a hybrid "Frankenstein" algorithm created in 2002 by a programmer named Tim Peters. It actually combines the best parts of **Merge Sort** and **Insertion Sort**.
    * **Use Case:** Real-world data is often already partially sorted. Timsort is specifically designed to look for those already-sorted chunks and take advantage of them, making it the perfect engine for Python's built-in sorting functions!

```python
# Example 1: Bubble Sort


# Unoptimised bubble sort
def bubble_sort(lst):
  # The outer loop keeps track of how many full passes we make through the list.
  for i in range(len(lst)):

    # The inner loop compares side-by-side items.
    # 'len(lst) - i' is a great trick: after each pass, the biggest numbers
    # bubble to the end. So, we don't need to check the last 'i' items again!
    for j in range(1, len(lst)-i):

      # Compare the current item (j) with the one right before it (j-1).
      # If the current one is smaller, they are in the wrong order.
      if lst[j] < lst[j-1]:

        # This is Python's clever shortcut to swap two items on a single line!
        lst[j], lst[j-1] = lst[j-1], lst[j]

  # Once all passes are done, return the fully sorted list.
  return lst

# Add the flag to optimise our algorithm
# This optimisation really makes a difference on lists that are partially sorted
def optimised_bubble_sort(lst):
  # The outer loop keeps track of how many full passes we make through the list.
  for i in range(len(lst)):

    # We use a 'flag' to check if any swaps happen during this pass.
    # If no swaps happen, we know the list is completely sorted!
    flag = False

    # The inner loop compares side-by-side items.
    # 'len(lst) - i' is a great trick: after each pass, the biggest numbers
    # bubble to the end. So, we don't need to check the last 'i' items again!
    for j in range(1, len(lst)-i):

      # Compare the current item (j) with the one right before it (j-1).
      # If the current one is smaller, they are in the wrong order.
      if lst[j] < lst[j-1]:

        # This is Python's clever shortcut to swap two items on a single line!
        lst[j], lst[j-1] = lst[j-1], lst[j]

        # Since we made a swap, we update our flag to True.
        flag = True

    # After checking the whole unsorted section, did we make any swaps?
    # If the flag is still False, the list is already in order. We can stop early!
    if flag == False:
      return lst

  # Once all passes are done, return the fully sorted list.
  return lst

print(bubble_sort(get_int_list(20))) # let's keep it fresh!
print(bubble_sort(get_str_list(10)))
```

```python
# Example 2: Insertion Sort

def insertion_sort(lst):
  # We start at index 1 (the second item), because a list with just one
  # item is already "sorted" by default!
  for i in range(1, len(lst)):

    # 'key' is the card we are currently holding and looking to place.
    key = lst[i]

    # 'j' points to the last card in the "already sorted" section of our hand.
    # It starts right before our current 'key' card.
    j = i - 1

    # We look to the left (j >= 0) AND check if the card we are looking at
    # is bigger than the 'key' we are holding.
    while j >= 0 and key < lst[j]:

      # If the card on the left is bigger, we shift it one spot to the right
      # to make room for our 'key'.
      lst[j + 1] = lst[j]

      # Move one more spot to the left to check the next card.
      j -= 1

    # Once we find a card smaller than our 'key' (or reach the very beginning),
    # we know we've found the perfect empty spot. We insert the 'key' card!
    lst[j + 1] = key

  # Return our newly sorted "hand" of cards
  return lst

# Let's test it out!
print(insertion_sort(get_int_list(20)))
print(insertion_sort(get_str_list(20)))
```

# Fast Sorting Algos

In both merge sort and quick sort, we'll see examples of the use of recursion.

```python
# Example 3: Merge Sort

# This is the "helper" function. Its only job is to take two ALREADY SORTED
# lists (a and b) and zip them together into one perfectly sorted list (k).
def merge(a, b):
  # i tracks our spot in list 'a', j tracks our spot in list 'b'.
  # k is the new, empty list where we will build our merged result.
  i, j, k = 0, 0, []

  # As long as neither list is empty, compare their front items!
  while i < len(a) and j < len(b):
    if a[i] < b[j]:
      k.append(a[i]) # 'a' has the smaller item, add it to k
      i += 1         # move our pointer in 'a' forward
    else:
      k.append(b[j]) # 'b' has the smaller item, add it to k
      j += 1         # move our pointer in 'b' forward

  # Once one of the lists is empty, we just scoop up whatever is left
  # in the other list and add it to the end of k.
  # (Using [i::] is a cool Python slice that means "from index i to the end")
  if i < len(a):
    k += a[i::]
  elif j < len(b):
    k += b[j::]

  return k

# This is the main function that splits the lists up!
def merge_sort(arr):
  # BASE CASE: If a list has 0 or 1 items, it can't be out of order!
  if len(arr) < 2:
    return arr

  # Find the middle index to chop the list in half
  mid = len(arr) // 2

  # Slice the array into a left half (a) and a right half (b)
  a = arr[:mid:]
  b = arr[mid::]

  # Here is the magic (Recursion!): We tell Merge Sort to sort the left half,
  # sort the right half, and then use our helper function to merge them back together.
  return merge(merge_sort(a), merge_sort(b))

# Try it out! (Assuming get_int_list and get_str_list are defined above in your notebook)
print(merge_sort(get_int_list(20)))
print(merge_sort(get_str_list(30)))
```

```python
# Example 5: Quick Sort

def quick_sort(arr):
  # BASE CASE: Just like merge sort, a list of 0 or 1 items is already sorted.
  if len(arr) < 2:
    return arr

  # Step 1: Pick a "pivot". We are just grabbing the item right in the middle.
  # You can pick any other item though. For example: pivot = arr[-1]
  mid = len(arr) // 2
  pivot = arr[mid]

  # Step 2: Divide! We use awesome Python List Comprehensions to filter the items.
  # Throw all items smaller than the pivot into a 'left' list.
  left = [i for i in arr if i < pivot]

  # Catch any items that are exactly EQUAL to the pivot in a 'middle' list.
  # (This is important just in case we have duplicate numbers in our list!)
  middle = [i for i in arr if i == pivot]

  # Throw all items bigger than the pivot into a 'right' list.
  right = [i for i in arr if i > pivot]

  # Step 3: Conquer! Sort the left side, sort the right side, and glue them
  # all together with the middle items packed in between!
  return quick_sort(left) + middle + quick_sort(right)

# Test it out
print(quick_sort(get_int_list(30)))
print(quick_sort(get_str_list(30)))
```
