````python
# Micro-challenge 1: Setting up the Student Database
!pip install faker
from faker import Faker
import uuid

fake = Faker()
names = sorted([fake.unique.name() for i in range(1000)])

def assign_uuids(namelist):
    """Return a nested list [[name, uuid_str], ...] using UUIDv4 (random)."""
    return [[name, str(uuid.uuid4())] for name in namelist]

namelist = assign_uuids(names)
print(f"Database created with {len(namelist)} students.")
print(f"First 10 names and ids: {namelist[:10]}")
````

### Task 2: Linear Search
Write a function `student_search` that performs a linear search to find a student's ID by their name.

````python
def student_search(namelist, target_name):
    for i in range(len(namelist)):
        if target_name == namelist[i][0]: # Retrieve the name using the 0th index
            return namelist[i][1] # Retrieve the id using the 0th index
    return -1 # Student does not exist

# Try it out:
print(student_search(namelist, names[0]))
````

### Task 3: Binary Search
Write a function `student_search_b` that performs a linear search to find a student's ID by their name.

````python
def student_search_b(namelist, target_name):
    low = 0
    high = len(namelist) - 1

    while low <= high:
        # Find the middle element
        mid = (low + high) // 2
        guess = namelist[mid][0]

        # Found student! Return its index.
        if guess == target_name:
            return namelist[mid][1]

        # The guess was too high; look at the left half
        elif guess > target_name:
            high = mid - 1

        # The guess was too low; look at the right half
        else:
            low = mid + 1

    # Student does not exist
    return -1

# Try it out:
print(student_search_b(namelist, names[0]))
````

### Task: The Versatile Search Function
Work with your partner to create a `search()` function that:
1. Asks the user for input.
2. If input is a number, search by price.
3. If input is 3 letters long, search by ID.
4. Otherwise, search by name.

Example Output:
```
Enter search term (or '0' to quit): 3.0

Results:
Name: Butter, ID: btt, Price: $3.0
Name: Milk, ID: mlk, Price: $3.0
------
```

*Hint: You may wanna google about try except in Python, it can be useful in this question :)*

````python
def search():
    """This function helps users search for a product in the supermarket by name, id or price."""
    while True:
        prompt = input("Enter search term (or '0' to quit): ")
        if prompt == '0':
            print("Goodbye!")
            break

        # We collect matching indices since multiple items can share a price
        matching_indices = []

        # 1. Try to see if the input is a number (Price Search)
        try:
            search_price = float(prompt)
            for index, p in enumerate(supermarket['price']):
                if p == search_price:
                    matching_indices.append(index)

        except ValueError:
            # If converting to float fails, it's text (ID or Name search)

            # 2. If input is exactly 3 letters long, search by ID
            if len(prompt) == 3:
                search_id = prompt.lower() # The ids are lower case
                if search_id in supermarket['id']:
                    matching_indices.append(supermarket['id'].index(search_id))

            # 3. Otherwise, search by Name
            else:
                search_name = prompt.lower() # Convert names to lower case
                for index, name in enumerate(supermarket['name']):
                    if name.lower() == search_name:
                        matching_indices.append(index)

        print("\nResults:")
        if matching_indices:
            for idx in matching_indices:
                print(f"Name: {supermarket['name'][idx]}, ID: {supermarket['id'][idx]}, Price: ${supermarket['price'][idx]:.1f}")
        else:
            print("No products found matching your search.")
        print("------")

search()
````

### Task 2: Phonebook Sorting
Implement a **bubble sort** function that sorts the phonebook by number.

````python
# Sample Phonebook
phonebook = [('Bob', '555-0103'), ('Charlie', '555-0102'), ('Alice', '555-0101')]

def phone_sort(phonebook):
    for i in range(len(phonebook) - 1):
        swapped = False # Optimised bubble sort
        for j in range(len(phonebook) - i - 1):
            if phonebook[j][1] > phonebook[j + 1][1]: # Compare numbers by retrieving index 1
                phonebook[j], phonebook[j + 1] = phonebook[j + 1], phonebook[j]
                swapped = True
        if swapped: # A complete pass of the phonebook without a swap means it is already sorted
            break

print(f'Unsorted phonebook: {phonebook}')
phone_sort(phonebook) # This function is in-place - it returns None and modifies the original list itself
# By contrast, out-of-place returns a new sorted list and leaves the original list unchanged

print(f'Sorted phonebook: {phonebook}')
````
