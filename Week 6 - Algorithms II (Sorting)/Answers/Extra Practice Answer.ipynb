**Q1. Linear Search (Last Occurance)**  
A list contains duplicate numbers. Write a function that returns the index of the last occurrence of target. For the list `A = [2, 1, 1, 6, 4, 2, 8, 2, 7, 9]`,  
- linear_search_dup(A, 2) returns 7
- linear_search_dup(A, 8) returns 6
- linear_search_dup(A, 5) returns -1

*Note: You may not use the list `index()` function.*

```python
def linear_search_dup(arr, target):
    # Start at the last index (len - 1) and move backwards to index 0
    for i in range(len(arr) - 1, -1, -1):
        if arr[i] == target:
            return i

    return -1

A = [2, 1, 1, 6, 4, 2, 8, 2, 7, 9]

print(linear_search_dup(A, 2))
print(linear_search_dup(A, 8))
print(linear_search_dup(A, 5))
```

**Q1a. Linear Search (All Occurances)**  
A list contains numbers with duplicates sorted in descending order. Implement linear search to return the index positions of all occurrences in a list. Include appropriate comments to explain your approach. For the list `A = [5, 3, 3, 3, 3, 3, 2, 2, 1]`,  

- `linear_search_dupall(A, 3)` returns `[1, 2, 3, 4, 5]`
- `linear_search_dupall(A, 1)` returns `[8]`
- `linear_search_dupall(A, 4)` returns `-1`

```python
def linear_search_dupall(arr, target):
    indices = []

    for i, num in enumerate(arr):
        if num == target:
            indices.append(i)

        # Optimization: since the list is sorted in descending order,
        # if the current number drops below the target, no more matches can exist.
        elif num < target:
            break

    # If the indices list has items, return it; otherwise, return -1 per instructions
    return indices if indices else -1

A = [5, 3, 3, 3, 3, 3, 2, 2, 1]

print(linear_search_dupall(A, 3))
print(linear_search_dupall(A, 1))
print(linear_search_dupall(A, 4))
```

`<method>_sort(lst, key, order)`  
**Description**:  
A recursive, stable sorting function that uses the Merge Sort algorithm to sort a list of elements based on a specific attribute and a defined direction.

Parameters:

- lst (list): The list of elements to be sorted (e.g., a list of integers, strings, or nested lists).

- key (function): A function (usually a lambda) that takes a single element from ls and returns the value to be used for comparison.

    - Example: lambda x: x[1] to sort by the second item in a sublist.

- order (string): Defines the sorting direction.

    - Accepts "ASC" for Ascending (smallest to largest).

    - Accepts "DESC" for Descending (largest to smallest).

**Critical Requirement**: Your sorting must be stable. If two elements have the same key, their relative order from the original list must not change.


---



The Multi-Pass Sort:
Use your merge_sort function to sort the following list of player data:
`players = [[80, 45], [95, 50], [80, 30], [95, 40], [70, 20]]`

Step A: Sort the list by Time_Taken (Index 1).

Step B: Take that resulting list and sort it again by Score (Index 0).

Result: Your correctly sorted list should look like `[[95, 40], [95, 50], [80, 30], [80, 45], [70, 20]]`

The Conclusion:
Why is it essential that your `<method>_sort` is stable for this two-step process to work? What would happen if it were unstable?

```python
def bubble_sort(lst, key, order):
    """
    A stable, in-place Bubble Sort function.
    Sorts a list based on a custom key function and specified order ('ASC' or 'DESC').
    """
    n = len(lst)

    for i in range(n - 1):
        # Track if a swap happened to optimize for already-sorted lists
        swapped = False

        # Last i elements are already in place
        for j in range(n - i - 1):
            val_current = key(lst[j])
            val_next = key(lst[j+1])

            # Determine if a swap is needed based on alignment
            should_swap = False
            if order == "ASC" and val_current > val_next:
                should_swap = True
            elif order == "DESC" and val_current < val_next:
                should_swap = True

            # STABILITY CRITICAL REQUIREMENT:
            # We ONLY swap if one is strictly greater/lesser than the other.
            # If val_current == val_next, we leave them alone to preserve original order.
            if should_swap:
                lst[j], lst[j+1] = lst[j+1], lst[j]
                swapped = True

        # If no two elements were swapped by the inner loop, the list is sorted!
        if not swapped:
            break

    return lst

players = [[80, 45], [95, 50], [80, 30], [95, 40], [70, 20]]

# Step A: Sort by Time_Taken (Index 1) in Ascending order (Faster is better)
bubble_sort(players, key=lambda x: x[1], order="ASC")
print("Step A (Sorted by Time):", players)

# Step B: Sort by Score (Index 0) in Descending order (Higher is better)
bubble_sort(players, key=lambda x: x[0], order="DESC")
print("Final Result (Sorted by Score):", players)
```
