````python
# Run this cell

def isbn(title: str, author: str, genre: str) -> str:
    """
    This algorithm returns a unique 13-digit ISBN code for a book.
    For simplicity's sake, we won't be using numbers.
    In real life, though, the ISBN-13 will be all numbers.
    """
    code = "978"
    code += title.strip().lower()[:4:]
    code += author.strip().lower()[:4:]
    code += genre.strip().lower()[:2:]

    return code

#testing the code
print(isbn("Sapiens", "Yuval Noah Harari", "History"))
````

## Part 1: Loading the Books 📂

Ms. Tan's data is in a CSV file. Your first job is to **read** the file and store the data as a **list of dictionaries** — one dictionary per book.

### What should each dictionary look like?
```python
{
    "title": "The Hobbit",
    "author": "J.R.R. Tolkien",
    "genre": "Fantasy",
    "copies": 3          # ← This should be an integer, not a string!
    "isbn": "978the j.r.fa" # Remember to generate the isbn and add it in!
}
```

### Your Task
Complete the function `load_books()` below.

**Hints:**
- Use `with open(...) as f:` and `csv.DictReader(f)` to read the file. `DictReader` automatically uses the first row as keys!
- Loop through each row and append it to a list. Remember to also include the isbn.
- Remember: everything from a CSV comes in as a **string**. You'll need to convert `copies` to an `int`.

````python
import csv

def load_books(filename):
    """
    Reads the CSV file and returns a list of book dictionaries.
    Each book has keys: title, author, genre, copies (int), ISBN.
    Remember to use the isbn() function above.
    """

    books = []
    with open(filename, newline="") as f:
        reader = csv.DictReader(f)          # first row becomes keys automatically
        for row in reader:
            book = {
                "title":  row["title"],
                "author": row["author"],
                "genre":  row["genre"],
                "copies": int(row["copies"]),               # string → int
                "isbn":   isbn(row["title"], row["author"], row["genre"]),
            }
            books.append(book)
    return books


# --- Test your function ---
books = load_books("library.csv")
print(f"Loaded {len(books)} books.")
print("First book:", books[0])
print("Type of copies:", type(books[0]["copies"]))  # Should print <class 'int'>

assert "978" in books[0]["isbn"] and len(books[-1]["isbn"]) == 13
````

✅ **Check:** You should see `Loaded 8 books.` and the type of copies should be `<class 'int'>`.

## Part 3: Searching for a Book 🔍

Ms. Tan wants students to be able to search for a book by **title** or **author**. The search should work even if the student types in **lowercase** or **partial words**.

For example, searching `"hobbit"` should find `"The Hobbit"`.

### Your Task
Complete the `search_books()` function.

**Hints:**
- Use `.lower()` on both the search term and the book fields so the case doesn't matter.
- Use the `in` operator to check if the search term appears anywhere in the title or author string.
- Collect all matching books into a list and return it.

````python
def search_books(books, search_term):
    """
    Returns a list of books where the search_term appears
    in the title OR the author (case-insensitive).
    """
    search_term_lower = search_term.strip().lower()
    results = []
    for book in books:
        if search_term_lower in book['title'].lower() or search_term_lower in book["author"].lower():
            results.append(book)

    return results


# --- Test your function ---
print("Searching for 'hobbit':")
found = search_books(books, "hobbit")
display_books(found)

print("Searching for 'Harari':")
found = search_books(books, "Harari")
display_books(found)

print("Searching for 'xyz' (should find nothing):")
found = search_books(books, "xyz")
if len(found) == 0:
    print("No books found.")
````

## Part 4: Borrowing a Book 📖

When a student borrows a book, the number of `copies` should **decrease by 1**. But if there are **0 copies left**, borrowing should not be allowed.

We'll search for the book by its **exact title** and update it.

### Your Task
Complete the `borrow_book()` function.

**Hints:**
- Loop through the books list. When you find the matching title:
  - If `copies > 0`: decrease copies by 1, print a success message, and `return` (stop the function).
  - If `copies == 0`: print a message saying it's unavailable, and `return`.
- If the loop finishes without finding the book, print "Book not found."

````python
def borrow_book(books, title):
    """
    Finds the book with the given title and reduces copies by 1.
    Prints a message to say if it was successful or not.
    """
    title_lower = title.strip().lower()
    for book in books:
        if book["title"].lower() == title_lower:
            if book["copies"] > 0:
                book["copies"] -= 1
                print(f"✅ You borrowed '{book['title']}'. Copies remaining: {book['copies']}")
                return
            else:
                print(f"❌ Sorry, '{book['title']}' has no copies available.")
                return
    print(f"❌ Book '{title}' not found in the library.")
    return


# --- Test your function ---
print("Borrowing 'A Brief History of Time' (1 copy):")
borrow_book(books, "A Brief History of Time")

print("\nBorrowing 'A Brief History of Time' again (should fail):")
borrow_book(books, "A Brief History of Time")

print("\nBorrowing a book that doesn't exist:")
borrow_book(books, "The Invisible Book")
````

## Part 5: Returning a Book 🔄

When a student returns a book, the number of `copies` should **increase by 1**.

````python
def return_book(books, title):
    """
    Finds the book with the given title and increases copies by 1.
    Prints a message to say if it was successful or not.
    """
    # TODO: Write this function yourself!
    # Steps:
    # 1. Loop through all books
    # 2. If the book's title matches:
    #    - Increase copies by 1
    #    - Print a success message
    #    - Return
    # 3. After the loop, print "Book not found."

    title_lower = title.strip().lower()
    for book in books:
        if book["title"].lower() == title_lower:
            book["copies"] += 1
            print(f"✅ '{book['title']}' returned. Copies now: {book['copies']}")
            return
    print(f"❌ Book '{title}' not found in the library.")
    return


# --- Test your function ---
print("Current copies of 'A Brief History of Time':", end=" ")
for b in books:
    if b["title"] == "A Brief History of Time":
        print(b["copies"])  # Should be 0 from Part 4

print("\nReturning 'A Brief History of Time':")
return_book(books, "A Brief History of Time")

print("\nCopies after return:", end=" ")
for b in books:
    if b["title"] == "A Brief History of Time":
        print(b["copies"])  # Should now be 1
````

## Part 6: Saving the Books 💾

After borrowing or returning books, we need to **save the updated data back to the CSV file**. Otherwise, all the changes will be lost when the program closes!

### Your Task
Complete the `save_books()` function.

**Hints:**
- Open the file in **write mode** (`"w"`) — this overwrites the old file.
- Use `csv.DictWriter` to write dictionaries.
- You need to define the column order (fieldnames): `["title", "author", "genre", "copies", "isbn"]`
- Call `writer.writeheader()` first to write the header row, then loop and call `writer.writerow(book)` for each book.

````python
def save_books(books, filename):
    """
    Writes the current list of books back to the CSV file.
    """
    with open(filename, "w", newline="") as f:
        writer = csv.DictWriter(f, fieldnames=["title", "author", "genre", "copies"])
        writer.writeheader()
        for book in books:
            writer.writerow({
                "title":  book["title"],
                "author": book["author"],
                "genre":  book["genre"],
                "copies": book["copies"],
            })
    print(f"💾 Library saved to '{filename}'.")

# --- Test your function ---
save_books(books, "library.csv")

# Reload and check the file was saved correctly
reloaded = load_books("library.csv")
print("\nReloaded from file. 'A Brief History of Time' copies:", end=" ")
for b in reloaded:
    if b["title"] == "A Brief History of Time":
        print(b["copies"])  # Should be 1
````

### Bonus 3: Save a Borrow Log

Every time a book is borrowed, **append a record** to a file called `borrow_log.csv` with the book title and the current copies remaining.

**Hint:** Open the file in **append mode** (`"a"`) so you don't overwrite previous records.

````python
import os

def log_borrow(title, copies_remaining):
    """
    Appends a borrow event to borrow_log.csv.
    Columns: title, copies_remaining
    """
    # TODO: Open 'borrow_log.csv' in append mode and write one row
    log_file = 'borrow_log.csv'

    file_is_new = not os.path.exists(log_file)
    with open(log_file, "a", newline="") as log:
        writer = csv.writer(log)
        if file_is_new:
            writer.writerow(["title", "copies_remaining"])
        writer.writerow([
            title,
            copies_remaining,
        ])
    print(f"📝 Borrow logged to '{log_file}'.")

# Now, we can modify borrow_book to utitlise log_borrow
def borrow_book(books, title):
    """
    Finds the book with the given title and reduces copies by 1.
    Prints a message to say if it was successful or not.
    """
    title_lower = title.strip().lower()
    for book in books:
        if book["title"].lower() == title_lower:
            if book["copies"] > 0:
                book["copies"] -= 1
                print(f"✅ You borrowed '{book['title']}'. Copies remaining: {book['copies']}")
                log_borrow(book['title'], book['copies']) # Additional line
                return
            else:
                print(f"❌ Sorry, '{book['title']}' has no copies available.")
                return
    print(f"❌ Book '{title}' not found in the library.")
    return


# Once done, call log_borrow() inside borrow_book() after a successful borrow.
print("Borrowing 'The Hobbit' (1 copy):")
borrow_book(books, "The Hobbit")
````
