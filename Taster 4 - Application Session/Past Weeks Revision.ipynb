```python
# recursion, week 4 cont'd

# a recursive function needs three parts: a function definition, a base case, and a part where it calls itself

#see below:

def fibonacci(n, a=0, b=1):
    if n == 0: # BASE CASE
        return a
    if n == 1: # BASE CASE
        return b

    return fibonacci(n - 1, b, a + b)

print(fibonacci(3))
```

```python
# args & kwargs

def regress(**kwargs):
    x,y = [],[]

    if "x" in kwargs:
        x = kwargs["x"]
    if "y" in kwargs:
        y = kwargs["y"]
    if "z" in kwargs:
        print("Ho ho ho, no such ability yet.")
        # does this line print? Under what circumstances?

    gradient = (y[0] - y[1]) / (x[0] - x[1])
    y_intercept = y[0] - (gradient * x[0])

    return f"The line equation is: y = {gradient}X + {y_intercept}"

print(regress(x=[1,2,3,4,5], y=[3,5,7,9,11]))
```
