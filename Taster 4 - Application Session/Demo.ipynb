## Project: Build a Quiz Game

We're going to build a quiz game that:
- Asks the player 5 questions
- Tracks their score
- Gives feedback (correct/wrong)
- Shows the final score at the end

Let's build this step by step!

---

### Step 1: Store Your Questions

First, we need to store our questions and answers. We'll use **variables** for this.

For now, let's store each question and answer as a separate variable:

```python
# Store our questions and answers as variables
question1 = "What is the capital of France?"
answer1 = "Paris"

question2 = "What is 5 + 3?"
answer2 = "8"

question3 = "What planet do we live on?"
answer3 = "Earth"

print("Questions stored! Ready to play.")
```

### Step 2: Ask One Question

Now let's ask the player ONE question and check if they got it right. We'll use **input()** to get their answer and **if/else** to check it.

```python
# Ask the player a question
player_answer = input(question1 + " ")

# Check if the answer is correct
if player_answer.lower() == answer1.lower():
    print("Correct! Well done!")
else:
    print(f"Sorry, the correct answer was: {answer1}")
```

### Step 3: Track the Score

We need a **variable** to keep track of the score. We'll start it at 0 and add 1 each time the player gets a correct answer.

```python
# Initialize the score
score = 0

# Ask the question
player_answer = input(question1 + " ")

# Check and update score
if player_answer.lower() == answer1.lower():
    print("Correct!")
    score = score + 1  # or: score += 1
else:
    print(f"Sorry, the correct answer was: {answer1}")

print(f"Your score so far: {score}")
```

### Step 4: Ask All Questions with a Loop

Instead of writing the same code 5 times, we can use a **loop**! We'll put our questions in a list and use a **for loop** to ask each one.

Note: We'll learn about lists in detail next week. For now, think of a list as a collection of items stored together.

```python
# Store questions and answers together as pairs
questions = [
    ("What is the capital of France?", "Paris"),
    ("What is 5 + 3?", "8"),
    ("What planet do we live on?", "Earth"),
    ("How many sides does a triangle have?", "3"),
    ("What is the largest mammal?", "Blue Whale")
]

# Initialize score
score = 0

# Loop through all questions
for question, answer in questions:
    player_answer = input(question + " ")
    
    if player_answer.lower() == answer.lower():
        print("Correct!")
        score += 1
    else:
        print(f"Sorry, the answer was: {answer}")
    
    print(f"Score: {score}\n")

print(f"=== GAME OVER ===")
print(f"Final Score: {score} / {len(questions)}")
```

### Step 5: Add a Play Again Feature

What if the player wants to try again? We can use a **while loop** to let them play multiple rounds. We'll ask if they want to play again after each round.

```python
# Full quiz game with replay feature
questions = [
    ("What is the capital of France?", "Paris"),
    ("What is 5 + 3?", "8"),
    ("What planet do we live on?", "Earth"),
    ("How many sides does a triangle have?", "3"),
    ("What is the largest mammal?", "Blue Whale")
]

playing = True

while playing:
    score = 0
    print("=== QUIZ TIME! ===\n")
    
    for question, answer in questions:
        player_answer = input(question + " ")
        
        if player_answer.lower() == answer.lower():
            print("Correct!")
            score += 1
        else:
            print(f"Sorry, the answer was: {answer}")
        
        print(f"Score: {score}\n")
    
    print(f"=== GAME OVER ===")
    print(f"Final Score: {score} / {len(questions)}")
    
    # Ask if they want to play again
    again = input("Play again? (yes/no): ")
    if again.lower() != "yes":
        playing = False
        print("Thanks for playing!")
```

## Debugging Tips

If your code isn't working:
1. **Read the error message** - it usually tells you the line number and what's wrong
2. **Use print()** - add print statements to see what your variables contain at each step
3. **Check your indentation** - Python is strict about spaces. Code inside if/else/for/while must be indented
4. **Check your colons** - every if, else, for, while line must end with a colon `:`
5. **Ask for help** - raise your hand, ask a tutor, ask a friend!
